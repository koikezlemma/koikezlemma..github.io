## 2-1. <font color="Teal">変数</font>

### 2-1-1. <font color="Teal">変数とは</font>

データを保持する際の容器に相当するものが変数です．  

<font color="OrangeRed">(っ´･ω･)っ</font><font color="IndianRed">【実はこの理解は Python においては若干不正確です．より正確にはすべてのデータは最初から容器に納められており，変数は単に容器の番号を覚えているだけです．このことは[後で](#scrollTo=2VulgXwbfK0X)少し問題になりますが，今は気にせずに進めていきましょう．】</font>

Pythonではひとつの変数にあらゆる型のデータを代入することができます．
文字列はダブルクォーテーション`" "`で囲うことにより，変数に代入できます．

<font color="RoyalBlue">【実習】 以下のように変数に値を代入したのち，出力する

*   変数 `price` に整数 1000 を代入
*   変数 `city` に文字列 Sendai を代入
</font>

```
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
price = 1000
city = "Sendai"

print(price)
print(city)
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
```

<font color="OrangeRed">(っ´･ω･)っ</font><font color="IndianRed">【他のプログラミング言語と同様に，イコール記号 `=` は数学の等号ではなく，代入を意味します．】</font>

### 2-1-2. <font color="Teal">画面への出力</font>

指定した文字列を画面に出力するには，上記のコードの通り， `print()`関数を使います．    
文字列以外を`print()`関数に渡すこともでき，その場合，[自動で文字列への変換が行われます](https://docs.python.org/ja/3/library/functions.html#print)．

また，`print()`関数では，文字列を出力する度に自動で改行を行います．

<font color="OrangeRed">(っ´･ω･)っ</font><font color="IndianRed">【自動で改行しないようにすることもできます．また，引数を複数にすることでふたつ以上の文字列を同時に出力することもできます．詳しくは [`print()`関数のドキュメント](https://docs.python.org/ja/3/library/functions.html#print) をご覧ください．】</font>

発展として，文字列の中に変数の値を埋め込みたいような場合，その解決法のひとつは`format()`メソッドを使用することです．

<font color="RoyalBlue">【実習】 `I bought a ... for ... yen.` の `...` の部分に `pen` と `100` を代入することで，`I bought a pen for 100 yen.` という文字列を作る．</font>

```
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
sentence = "I bought a {0} for {1} yen."
print(sentence.format("pen", 100))
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
```

出力したいベースの文字列の中で，後から値を代入したい部分に`{0}, {1}, ...` という風に名前をつけておき，`format()` メソッドの中でそれらの値を順に指定します．なお，`{ }` の中の番号を省略すると，自動で0から順に番号がつけられます．

<font color="OrangeRed">(っ´･ω･)っ</font><font color="IndianRed">【`format()` メソッドを使うと，出力される値の表示形式を指定したりもできます．`format()` メソッドの更なる使い方については，Pythonドキュメントの[str.format()構文](https://docs.python.org/ja/3/library/stdtypes.html#str.format)や，[書式指定例](https://docs.python.org/ja/3/library/string.html#format-examples)をご覧ください．また，文字列に変数の値を埋め込む他の方法として，[フォーマット済み文字列リテラル](https://docs.python.org/ja/3/tutorial/inputoutput.html?highlight=format#formatted-string-literals)という方法もあります．】</font>

### 2-1-3. <font color="Teal">変数名</font>

変数名にはアルファベット，数字，アンダースコア（_）などの文字がよく使用されます．  
アルファベットの大文字と小文字は別の文字として認識されます．

<font color="OrangeRed">(っ´･ω･)っ</font><font color="IndianRed">【変数名に日本語なども使えますが一般的とはいえません．[PEP8の命名規約](https://pep8-ja.readthedocs.io/ja/latest/#ascii)では標準ライブラリで使われる識別子はASCIIと互換性がなければならないと定められています（つまり，標準ライブラリを開発する際は変数名等には半角のアルファベット，数字，記号しか使えないということです）．】</font>

また以下のようなことは禁止事項です．

*   変数の最初の文字が数字であること．（例えば， `1st_trial` というような変数名は禁止）
*   [予約語](https://docs.python.org/ja/3/reference/lexical_analysis.html#keywords)と等しい文字列．例えば，変数名を `if` や `as` にすることはできません．


<font color="OrangeRed">(っ´･ω･)っ</font><font color="IndianRed">【予約語以外であっても他で使用している名前は使わない方がいいです．例えば，変数名を`print` としてそこに 5 を代入してしまうと，出力関数としての `print` が使用できなくなってしまいます．ちなみに，もしうっかりこのような処理をしてしまったらブラウザ画面上部の「ランタイム」から「ランタイムを出荷時設定にリセット」を選択し，設定をリセットしましょう．】</font>

アルファベットを大文字にする際のルールを決めておくと可読性が向上します．
これらの流儀については[PEP8の命名規則](https://pep8-ja.readthedocs.io/ja/latest/#id24)を参考にしてください．

この教材では，変数名には英語の小文字と数字，アンダースコアを利用します．

### 2-1-4. <font color="Teal">複数変数への同時代入</font>

Python では複数の変数に一気に値を代入することができます．

<font color="RoyalBlue">【実習】 変数 `a, b, c` に一気に値を代入してみる．</font>

```
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
a, b, c = 1, 2, "Hello"
print(a)
print(b)
print(c)
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
```

<font color="OrangeRed">(っ´･ω･)っ</font><font color="IndianRed">【複数変数への同時代入は，単にタプルというデータ構造への代入処理をしているに過ぎません．詳細は[Pythonのチュートリアル](https://docs.python.org/ja/3/tutorial/datastructures.html#tuples-and-sequences)をご覧ください．】</font>

## 2-2. <font color="Teal">数値</font>

Pythonでは，[整数，浮動小数点数，複素数の3種類の数値を扱えます](https://docs.python.org/ja/3/library/stdtypes.html#numeric-types-int-float-complex)．

複素数については本教材では省略します．

整数は小数点以下の値を含まない数値を表すために使用し，浮動小数点数は小数点以下の値を含む数値を表すために使用します．

数値を記述した際，そこに小数点を含まなければ整数とみなされ，小数点を含むと浮動小数点とみなされます．  
例えば，2， -1， 999 と書くと整数とみなされ，1.2， 2.0， -10.5 と書くと浮動小数点数とみなされます．

<font color="OrangeRed">(っ´･ω･)っ</font><font color="IndianRed">【2 という数値を扱う際，2と記述すると整数扱いとなり，2.0と記述すると浮動小数点数扱いとなるので，書き方に注意しましょう．】</font>

### 2-2-1. <font color="Teal">算術演算</font>

Pythonでも加減乗除の計算が可能です．

掛け算を行うためには `*` を使用します．

割り算には `//` と `/` の2つの演算子があります．

演算子 `//` は整数の割り算に用いる記号で，整数の割り算の商を求めます．例えば $5 \div 2 = 2$ 余り 1 となりますが，`5 // 2` により商の 2 を求めることができます．

演算子 `/`  は商を小数点以下まで計算するものです．$5 \div 2 = 2.5$ なので，`5 / 2` は2.5になります．`/` を使って得られる値は常に浮動小数点数であることに注意が必要です．例えば，`4 / 2`とすると浮動小数点の数 2.0 が得られます．

<font color="RoyalBlue">【実習】 整数の加減乗除を試してみる．</font>

```
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
print(2 + 3)
print(4 - 1)
print(2 * 3)
print(5 // 2)
print(5 / 2)
print(4 / 2)
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
```

算術演算において整数と浮動小数点数が混在している場合，計算結果は浮動小数点数となります．整数同士の演算結果は基本的には整数ですが，上記で述べたとおり，`/` を使って得られる値は常に浮動小数点数になります．

<font color="OrangeRed">(っ´･ω･)っ</font><font color="IndianRed">【C言語等の他の言語では，整数同士の演算の結果は必ず整数になるので，それらの言語を使用している人にとっては，整数 / 整数 が浮動小数点数になるというのは大きな違和感を感じるかもしれません．】</font>

<font color="OrangeRed">(っ´･ω･)っ</font><font color="IndianRed">【浮動小数点数の割り算に `//` を使用することも一応可能で，その場合は計算結果から小数点以下の値を切り捨てた値（浮動小数点数）が得られますが，小数点以下を切り捨てた整数値を得たい場合には，`/` を使って売られた結果に対し [math.floor()](https://docs.python.org/ja/3/library/math.html#math.floor) を使うのが一般的と思われます．】</font>

<font color="RoyalBlue">【実習】 浮動小数点数を含む場合の加減乗除を試してみる．</font>

```
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
print(1.1 + 0.1)
print(2 - 0.1)
print(0.1 * 3)
print(3.0 * 4)
print(8.0 / 2.0)
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
```

<font color="OrangeRed">(っ´･ω･)っ</font><font color="IndianRed">【計算の結果，ものすごく桁数の長い数値が得られているかもしれません．これは計算の際に生じた誤差によるものです．浮動小数点数を使用する際は，このような計算誤差が生じることを踏まえておく必要があります．】</font>

整数の割り算をした際の余りは `%` によって取得できます．例えば，$5 \div 2 = 2$ 余り 1 なので，`5 % 2` は 1 になります．

<font color="RoyalBlue">【実習】 整数の割り算 $5 \div 2$ の余りが 1 であることを確認する．</font>

```
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
5 % 2
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
```

べき乗を計算することも可能です．
$3^4$ を計算するには `3 ** 4` と書きます．

<font color="RoyalBlue">【実習】 $3^4 = 81$ を確認する．</font>

```
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
3 ** 4
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
```

### 2-2-2. <font color="Teal">変数の値の更新</font>

同様に， `-=`， `*=`， `/=`, `//=` なども利用可能です．

<font color="RoyalBlue">【実習】 `+=` を使用して変数 `a` の値を 10 から 5 だけ増加させてみる．</font>

```
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
a = 10
a += 5
a
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
```

### 2-2-3. <font color="Teal">文字列から数値への変換</font>

文字列として書かれている数字を数値として扱いたい場合には，変換処理が必要となります．文字列を整数に変換するには [int()](https://docs.python.org/ja/3/library/functions.html#int) を使用します．

<font color="OrangeRed">(っ´･ω･)っ</font><font color="IndianRed">【浮動小数点数への変換には[float()](https://docs.python.org/ja/3/library/functions.html#float)という関数を使うことができます．】</font>

## 2-3. <font color="Teal">リスト</font>

### 2-3-1. <font color="Teal">リストとは</font>

リストとは，複数個のデータをまとめて管理するためのデータ構造のひとつです．  
リストには，数値，文字列等の様々な種類のデータを格納することができます．  
 `[2, 3, 5]` と書くと 2, 3, 5 の 3 つの要素からなるリストを表します．

<font color="OrangeRed">(っ´･ω･)っ</font><font color="IndianRed">【リストには要素ごとに種類の異なるデータを格納することもできますが，そのようにすると可読性が下がることがあるので気をつけましょう．[リストのすべての要素を同じデータ型にする方が一般的です](https://docs.python.org/ja/3/tutorial/introduction.html?highlight=%E9%80%9A%E5%B8%B8%E3%81%AF%E5%90%8C%E3%81%98%E5%9E%8B%E3%81%AE%E8%A6%81%E7%B4%A0%E3%81%AE%E3%81%BF%E3%82%92%E6%8C%81%E3%81%A1%E3%81%BE%E3%81%99#lists)．】</font>

<font color="RoyalBlue">【実習】 "Chinese", "English", "Indonesian", "Japanese" の 4つの文字列からなるリストを作成し，変数 `languages` に格納する．</font>

```
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
languages = ["Chinese", "English", "Indonesian", "Japanese"]
print(languages)
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
```

`print()` を使ってリストを出力すると，リストのすべての要素が出力されます．

### 2-3-2. <font color="Teal">リスト要素へのアクセス</font>

リストでは，格納されている各要素に番号がつけられており，それをインデックスと呼びます．  
先頭要素には 0，その次の要素には 1 という感じで， 0から順に整数の番号がつけられています．

<font color="OrangeRed">(っ´･ω･)っ</font><font color="IndianRed">【先頭要素のインデックスを 0 にするか，1にするかについては，プログラミング言語ごとに仕様が分かれていますが，Python を含む多くの言語では，先頭要素のインデックスを 0 にしています．これは，慣れないと少し違和感があるかもしれません．日常生活では番号付けをする際，先頭要素の番号を 1 にすることが多いように思います．例えば学校のクラスにおける出席番号が 0 から始まることは少なく，1 から始まることが多いと思われます．】</font>

インデックスを指定することでリストの要素を取得することができます．

<font color="RoyalBlue">【実習】 上記のリストにおいて先頭から2番目の要素を取得して表示する．</font>

```
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
languages = ["Chinese", "English", "Indonesian", "Japanese"]
languages[1]
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
```

<font color="OrangeRed">(っ´･ω･)っ</font><font color="IndianRed">【先頭から2番目の要素のインデックスは 1 であることに気をつけてください．】</font>

### 2-3-3. <font color="Teal">負のインデックス</font>

インデックスを負にすることで，後ろから数えた際の順番を指定することができます．  
最終要素は -1，後ろから2番目の要素は -2 （以下同様）です．

<font color="RoyalBlue">【実習】 上記のリストにおいて，最終要素を表示する．</font>

```
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
languages = ["Chinese", "English", "Indonesian", "Japanese"]
languages[-1]
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
```

### 2-3-4. <font color="Teal">要素の追加と削除</font>

リストの末尾に新しい要素を追加するには，`append()` メソッドを使います．

<font color="RoyalBlue">【実習】 上記リストの末尾に "German" を追加する．</font>

```
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
languages = ["Chinese", "English", "Indonesian", "Japanese"]
languages.append("German")
languages
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
```

<font color="OrangeRed">(っ´･ω･)っ</font><font color="IndianRed">【リストの末尾以外の場所に要素を追加する場合はメソッド `insert()` を使用します．詳しくは
[`insert()`のドキュメント](https://docs.python.org/ja/3/tutorial/datastructures.html?highlight=insert)を参照してください．】</font>

`pop()` メソッドを使用すると，リストの最終要素を取得し，その要素をリストから削除することができます．
リストから要素をひとつずつ取り出して処理するような場合に使用されます．

<font color="RoyalBlue">【実習】 上記リストから末尾の要素を取得し，その要素をリストから削除する．</font>

```
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
languages = ["Chinese", "English", "Indonesian", "Japanese", "German"]

lang = languages.pop()
print(lang)
print(languages)
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
```

<font color="OrangeRed">(っ´･ω･)っ</font><font color="IndianRed">【プログラミングを行う際の最も基本的なデータ構造の１つにスタックがありますが，`append()`と`pop()`を用いることで，簡単にスタックを実装できます．】</font>

<font color="OrangeRed">(っ´･ω･)っ</font><font color="IndianRed">【リストの最終要素でなく，特定インデックスの要素を削除する場合にも [`pop()` メソッド](https://docs.python.org/ja/3/tutorial/datastructures.html?highlight=pop)を使用しますが，その際は括弧の中に削除したいインデックスを指定します．】</font>

<font color="OrangeRed">(っ´･ω･)っ</font><font color="IndianRed">【指定した場所の要素を削除するのではなく，指定した値の要素を削除する場合は `remove()` メソッドを使用することができます．しかし，指定した要素が見つからない場合，エラーが発生するので要注意です．また，指定した要素が複数見つかった時はインデックスが一番小さい要素のみが削除されます．リストから条件を満たす要素をすべて削除する高速で安全な方法として，Python公式ドキュメントの[FAQ](https://docs.python.org/ja/3/faq/programming.html?highlight=remove#how-do-you-remove-multiple-items-from-a-list)では `mylist[:] = [x for x in mylist if keep_condition]` という書き方が紹介されています．詳細説明は省略しますが，上記のリストから "English" という値の要素をすべて削除する場合は，`languages[:] = [x for x in languages if x != "English"]` と書けます．】</font>

### 2-3-9. <font color="Teal">リストの並べ替え</font>

リストに対して並び替えを行うには [`sort()` メソッド](https://docs.python.org/ja/3/library/stdtypes.html?highlight=sort#list.sort)を使用します．

<font color="RoyalBlue">【実習】 リスト `[4, 2, 3, -4, 8, 0, 9]` を昇順に（小さいものほど前に来るように）並び替える．</font>

<font color="RoyalBlue">【実習】 リスト `[4, 2, 3, -4, 8, 0, 9]` を降順に（大きいものほど前に来るように）並び替える．</font>

```
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
numbers = [4, 2, 3, -4, 8, 0, 9]
numbers.sort(reverse=True)
numbers
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
```

<font color="RoyalBlue">【実習】 リスト `["French", "Indonesian", "English", "Chinese"]` を辞書式順序で（各要素の文字列を辞書に掲載することを考えた場合に，前のページに出てくる要素ほどリストの前に来るように）並び替える．</font>

```
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
languages = ["French", "Indonesian", "English", "Chinese"]
languages.sort()
languages
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
```

<font color="OrangeRed">(っ´･ω･)っ</font><font color="IndianRed">【2つのアルファベット文字列の辞書式順序を求めるには，まず先頭文字のアルファベット順の比較で順序を決めようとし，もし同じアルファベットだったら2番目の文字のアルファベット順で順序を決める，以下同様というような手順を行えば良いです．】</font>

<font color="OrangeRed">(っ´･ω･)っ</font><font color="IndianRed">【与えられたリストはそのままにしておいて，それとは別にソートされたリストを生成する場合には，[`sorted()` メソッド](https://docs.python.org/ja/3/library/functions.html?highlight=sorted#sorted)を利用します．】</font>



### 2-3-10. <font color="Teal">スライス</font>

リストからその一部（または全部）を抜き出したものを「スライス」と呼びます．  
スライスを使うことで，リストから複数要素を一気に抜き出すことができます．


<font color="OrangeRed">(っ´･ω･)っ</font><font color="IndianRed">【リストから一部を抜き出すと言っても，元のリストの要素を削除するわけではありません．元のリストの一部を要素とする新たなリストを作ることをこう呼んでいます．】</font>

例として，リスト 
```
foods = ["サラダ", "ハンバーガー", "ポテト", "ナゲット"]
```
から一部の要素を抜き出して新しいリストを作ることを考えます．

まず，基本的なやり方として，`foods[1:3]` と書くと，インデックス 1 以上，インデックス 3 未満のすべての要素を抜き出してリストを生成します．よってこの記法により生成されるリストは `["ハンバーガー", "ポテト"]` となります．`foods[1:3]` と書いた時，1 を開始インデックス，3 を終了インデックスと呼ぶことにします．

<font color="OrangeRed">(っ´･ω･)っ</font><font color="IndianRed">【`foods[1:3]` と書いた時，インデックス 1 の要素は含まれるが，インデックス 3 の要素は含まれないことに注意しましょう．つまりこの書き方ではインデックスが 1 以上 2 以下の要素が抜き出されます．】</font>

<font color="RoyalBlue">【実習】 リスト `foods = ["サラダ", "ハンバーガー", "ポテト", "ナゲット"]` からインデックスが 1 以上 3 未満のすべての要素を抜き出してリストを生成する．</font>

```
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
foods = ["サラダ", "ハンバーガー", "ポテト", "ナゲット"]
foods[1:3]
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
```

次に，インデックス 1 から最終要素までを抜き出す場合は `foods[1:]` という風に終了インデックスを書かないようにします．

<font color="RoyalBlue">【実習】 上記リストからインデックス 1 以上のすべての要素を抜き出してリストを生成する（`["ハンバーガー", "ポテト", "ナゲット"]` が生成される）．</font>

```
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
foods = ["サラダ", "ハンバーガー", "ポテト", "ナゲット"]
foods[1:]
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
```

次に，先頭要素からインデックス 3 未満の要素までを抜き出す場合には， `foods[0:3]` と書くか，もしくは，開始インデックスを省略して， `foods[:3]` と書きます．

<font color="RoyalBlue">【実習】 上記リストからインデックス 3 未満のすべての要素を抜き出してリストを生成する（`["サラダ", "ハンバーガー", "ポテト"]` が生成される）．</font>

```
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
foods = ["サラダ", "ハンバーガー", "ポテト", "ナゲット"]
foods[:3]
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
```

開始インデックスと終了インデックスには負の値を指定することもできます．これはインデックスによるアクセスの時と同じく，後ろから数えた場合の順番を表すものです．最後尾の時 -1 ，後ろから 2 番目の時 -2 になります．

<font color="RoyalBlue">【実習】 上記リストの後ろから  2 番目から最終要素までの要素を抜き出してリストを生成する．（`["ポテト", "ナゲット"]` が生成される）．</font>

```
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
foods = ["サラダ", "ハンバーガー", "ポテト", "ナゲット"]
foods[-2:]
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
```

スライスの拡張として，指定範囲の要素を一定間隔で抜き出してリストを生成することもできます．  
`a[i:j:k]`のように従来のスライスに`:k`を追加すると，リスト`a`からインデックス`i`以上`j`未満の要素を`k`間隔で取り出します．

<font color="RoyalBlue">【実習】 上記リスト`foods`からインデックス 3 未満の要素を一つ飛ばしで抜き出してリストを生成する（`["サラダ", "ポテト"]` が生成される）．</font>

```
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
foods = ["サラダ", "ハンバーガー", "ポテト", "ナゲット"]
foods[:3:2]
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
```

### 2-3-11. <font color="Teal">リストの複製</font>

リストを複製するために `=` を使うことはできません．  
つまり，`list_a = list_b` と書いても `list_a` に `list_b` の内容が複製されることはありません．

リストを複製して新たなリストを生成するための方法として[`copy()`](https://docs.python.org/ja/3/tutorial/datastructures.html?highlight=copy)メソッドがあります．



<font color="RoyalBlue">【実習】 上記リストを複製して新しいリスト `menu` を作る．</font>

```
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
foods = ["サラダ", "ハンバーガー", "ポテト", "ナゲット"]
menu = foods.copy()
menu
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
```

<font color="OrangeRed">(っ´･ω･)っ</font><font color="IndianRed">【`foods.copy()`と書く代わりに`foods[:]` と書いても`foods` の複製が生成されます．ただ，この方法は機械学習等でもよく使われているNumPy配列では使えないので，ここでは，NumPy配列でも使える `copy()` の方を紹介しました．】</font>

<font color="OrangeRed">(っ´･ω･)っ</font><font color="IndianRed">【リストが入れ子になっているとき（リストの要素がリストになっているような時），`copy()` メソッドを使った方法（および`foods[:]`を使った方法）はうまくいきません．代わりに，copyモジュールをインポート（`import copy`）して,`menu = copy.deepcopy(foods)` のように書く必要があります．詳しくは[`deepcopy()`のドキュメント](https://docs.python.org/ja/3/library/copy.html?highlight=deepcopy#copy.deepcopy)をご覧ください．】</font>

### 2-3-12. <font color="Teal">【参考】リスト複製詳細</font>

ここからは，`=` を使ってリストの複製ができないことの詳細について説明したいと思います．

`list_a = list_b` と書くと，`list_b` の複製が `list_a`に格納されるわけではなく，単に `list_a` が `list_b` の別名となります．
つまり，１つのリストに対して，`list_a` という名前と `list_b` という名前の 2 つが存在することになり，どちらの名前を使っても同一のリストを操作することになります．
これは例えていうのならば，Windows OSにおいて，ファイルの複製を作るつもりが，誤ってファイルのショートカット（Linuxであればシンボリックリンク）を作ってしまったという状況に似ています．
ショートカットのファイルを開くと元のファイルが開かれることになります．

複製でなく別名になってしまうとどのようなまずいことが起こるのかということについて，コードを書いて試してみたいと思います．

以下のようなシナリオを考えてみます．ハンバーガーショップで私と友人が注文をするとして，私がハンバーガーとポテト，友人が私と同じ注文に加えサラダを追加注文するとします．

この時，私の注文リストを友人に複製したのち，友人の注文にサラダを追加すれば良いですが，注文リストを複製しようとして誤ってリストの代入を使ってしまうとまずい事態になります．

<font color="RoyalBlue">【実習】 上記のシナリオを実装してみる．</font>

```
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
my_order = ["ハンバーガー", "ポテト"]
friend_order = my_order

friend_order.append("サラダ")

print("My order is\t\t: {0}".format(my_order))
print("My friend's order is\t: {0}".format(friend_order))
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
```

注文リストをみると，私の注文リストに頼んでもいないサラダが入っています．このように，リストの代入を使うとまずい事態になってしまいます．

さて，上記を正しい処理にするには，友人の注文リスト `friend_order` に `my_order` を代入するのではなく，`my_order.copy()` （`my_order` の複製）を代入するようにすれば良いです．

<font color="RoyalBlue">【実習】 上記のシナリオについて正しい実装に修正する．</font>

```
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
my_order = ["ハンバーガー", "ポテト"]
friend_order = my_order.copy()

friend_order.append("サラダ")

print("My order is\t\t: {0}".format(my_order))
print("My friend's order is\t: {0}".format(friend_order))
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
```

無事，私の注文リストからサラダがなくなりました．

<font color="OrangeRed">(っ´･ω･)っ</font><font color="IndianRed">【このようなことが起こってしまう原因をきちんと理解するには，Python のオブジェクトについて知る必要があります．Python では数値やリストなどのデータはすべてオブジェクトという単位で管理されています．変数に値やリストを代入するという行為は変数にオブジェクトのIDを覚えさせるという行為にほかなりません．そして，`list_a = list_b` と書くと`list_a` と `list_b` が保持するオブジェクトIDが等しくなってしまいます．これが上記の事態が発生する原因です．なお，数値の代入の場合も同じ問題が起こるのではないかという疑問が出てきますが，数値（より一般的にはイミュータブルなオブジェクト）の場合には値を更新するたびにオブジェクトIDが変化するため，このような問題は起こりません．これについての詳細はPython公式ドキュメントの[3. データモデル](https://docs.python.org/ja/3/reference/datamodel.html#objects-values-and-types)をご覧ください．】</font>

## 2-4. <font color="Teal">タプル</font>

### 2-4-1. <font color="Teal">タプルとは</font>

タプルは複数の決まった数の要素をまとめて管理するためのデータ構造です．  
要素数を増やしたり，要素の値を個別に更新したりできないことが特徴です．

リストの生成には角括弧 `[ ]` を使用しましたが，タプルの生成には丸括弧 `( )` を使用します．ただし，丸括弧は多くの場合に省略可能です．  

リストと同様にインデックスやスライスを用いたアクセスができます．

<font color="OrangeRed">(っ´･ω･)っ</font><font color="IndianRed">【リストとタプルの大きな違いは，リストがリスト内の各要素の値を自由に変更できるのに対し，タプルの各要素は個別に値を変更できない点にあります（厳密にはオブジェクトIDの値を変えないまま値を変更することができないということです）．タプルのこのような特徴を「イミュータブル」と呼びます（リストのように中身の値を変更できるオブジェクトは「ミュータブル」であると言われます）．】</font>

<font color="RoyalBlue">【実習】 x,y,z 座標を保持するタプルを生成し，各座標値を出力する．</font>

```
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
xyz_coord = (2, -1, 3)
print(xyz_coord[0])
print(xyz_coord[1])
print(xyz_coord[2])
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
```

<font color="RoyalBlue">【実習】 タプルの各要素の値を複数の変数に同時代入できることを確認する．</font>

```
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
xyz_coord = (2, -1, 3)
(x, y, z) = xyz_coord
print(x)
print(y)
print(z)
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
```

<font color="RoyalBlue">【実習】 上記の処理でタプルの丸括弧を省略できることを確認する．</font>

```
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
xyz_coord = 2, -1, 3
x, y, z = xyz_coord
print(x)
print(y)
print(z)
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
```

<font color="OrangeRed">(っ´･ω･)っ</font><font color="IndianRed">【これをみると，複数変数への同時代入が，実はタプルの代入処理だということがよく変わると思います．】</font>

<font color="RoyalBlue">【実習】 タプルの要素の値を変更しようとするとエラーが発生することを確認する．</font>

```
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
xyz_coord = (2, -1, 3)
xyz_coord[0] = 5
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
```

<font color="RoyalBlue">【実習】 タプル自体を作り直すことで，タプルの特定要素の値を変更する．</font>

```
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
xyz_coord = (2, -1, 3)
print("Original tuple\t: ", xyz_coord)
xyz_coord = (5, xyz_coord[1], xyz_coord[2])
print("New tuple\t: ", xyz_coord)
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
```

### 2-4-2. <font color="Teal">1要素からなるタプル</font>

タプルの要素数を1にすることもできますが，この場合，書き方に注意が必要です．  
100という整数値1つからなるタプルを作ろうとして `(100)` と書くと，タプルでなく整数 `100` と解釈されてしまいます．  
タプルを作るには，要素の後ろにカンマをつけて，`(100,)` のように書きます．

<font color="RoyalBlue">【実習】 z座標 `3`を保持するタプル `z_coord` を作成する．</font>

```
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
z_coord = (3,)
print(z_coord)
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
```

このセクションでは，辞書（ディクショナリ）と呼ばれるデータ型を学びます．辞書はリストと同じく複数のデータを保持するためのものですが，リストではインデックスを用いて要素にアクセスするのに対し，辞書ではキーと呼ばれる値を使って要素にアクセスします．辞書はデータへの高速アクセスという観点からもとてもよく用いられるデータ構造です．

## 2-5. <font color="Teal">辞書</font>

### 2-5-1. <font color="Teal">辞書とは</font>

辞書もリスト等と同じく複数の要素をまとめて管理できるデータ構造です．

辞書では，保持したい値（value）をキー（key）と呼ばれる値とペアにして格納します．
キーは各要素に付けられた名前を表します．
辞書では，要素を指定する際，インデックスではなくキーを指定します．

リストは `[ ]` を使って定義され，タプルは `( )` を使って定義されましたが，辞書は `{ }` を使って定義されます．  
複数の料理の値段を格納したい時，辞書を使うと以下のように書けます．
```
prices = {"pasta": 700, "pizza": 600, "ramen": 600}
```
この時，料理名がキーで価格が値となります．  
`"pasta"` が700（円），`"pizza"` が600（円），`"ramen"` が600（円）です．  
上記のようにキーと値はコロン `:`で区切ります．

この辞書から`"pasta"` の値段を取得するには，`prices["pasta"]`と書きます．この時，`"pasta"` がキーで `700` が値です．

各要素にキーが紐づいていることによって，インデックスを用いてアクセスする場合と比較して，所望のデータへのアクセスが容易になっていることがわかると思います．

<font color="RoyalBlue">【実習】 `"pasta"` が700（円），`"pizza"` が600（円），`"ramen"` が600（円）である．この情報を辞書 `prices` に格納したのち，辞書から "pasta" の価格を取得する．</font>

```
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
prices = {"pasta": 700, "pizza": 600, "ramen": 600}
print(prices["pasta"])
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
```

### 2-5-2. <font color="Teal">新たな要素の追加</font>

辞書 `prices` に `new_key, new_value` のペアを追加するには `prices[new_key] = new_value` のように書きます．

<font color="RoyalBlue">【実習】 上記の辞書 `prices` に"soba" が 400（円）という情報を追加する．</font>

```
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
prices = {"pasta": 700, "pizza": 600, "ramen": 600}
prices["soba"] = 400
print(prices)
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
```

<font color="OrangeRed">(っ´･ω･)っ</font><font color="IndianRed">【辞書を`print()`文で出力すると，すべてのキーに対する値が出力されます．】</font>

空の辞書（要素が１つも格納されていない辞書）を作るには `{ }` と書きます．

<font color="RoyalBlue">【実習】 空の辞書 `prices` を作成した後，"pasta"が700（円），"soba" が 400（円）という情報を追加する．</font>

```
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
prices = {}
prices["pasta"] = 700
prices["soba"] = 400
prices
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
```

### 2-5-3. <font color="Teal">値の変更</font>

辞書の要素の値を変更するのは，リストと同様に，指定した要素に新しい値を代入します．

<font color="RoyalBlue">【実習】 上記の辞書 `prices` において，"soba"の価格を500（円）に変更する．</font>

```
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
prices = {"pasta": 700, "soba": 400}
prices["soba"] = 500
prices
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
```

### 2-5-4. <font color="Teal">要素の削除</font>

指定したキーとそれに対応する値を辞書から削除するには `pop()` メソッドを使用します．  
`pop()` の括弧の中にはキーを指定します．  
`pop()` メソッドを呼ぶと指定したキーに対応する値を取得できますが，同時にそのキーと値のペアが辞書から削除されます．

<font color="RoyalBlue">【実習】 上記の辞書 `prices` から "soba"の情報を削除する．</font>

```
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
prices = {"pasta": 700, "soba": 500}
prices.pop("soba")
prices
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
```

### 2-5-5. <font color="Teal">キーの存在確認</font>

指定したキーが辞書に存在するかどうかをチェックするには，`キー in 辞書` のように書きます．  
キーが存在すれば `True`，存在しなければ `False` になります．

<font color="RoyalBlue">【実習】 "pasta"が700（円），"soba" が 500（円）という情報が格納された辞書 `prices` を作成した後，"pasta"と"sushi"の2つのキーが辞書に含まれているかをチェックする．</font>

```
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
prices = {"pasta": 700, "soba": 500}
print("pasta" in prices)
print("sushi" in prices)
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
```

### 2-5-6. <font color="Teal">集合</font>

辞書に似たデータ構造として集合があります．これは数学の集合を扱うことができます．  
集合を定義する際は，辞書と同じく `{ }` を使用します．

<font color="RoyalBlue">【実習】 料理を要素とする集合を作成する．</font>

```
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
food_set = {"crepe", "salad", "tsunamayo", "yakitori"}
food_set
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
```

<font color="OrangeRed">(っ´･ω･)っ</font><font color="IndianRed">【辞書と同様に空の集合を作ることもできますが，空の集合を作ろうとして `{ }` と書いてしまうと，空の集合でなく空の辞書が作られてしまうので注意してください．空の集合を作る時は `food_set = set()` のように記載します．】</font>

和集合を求めたり，差集合を求めたりすることができます．

<font color="RoyalBlue">【実習】 既存の料理集合と新しい料理集合の和集合を求める．</font>

```
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
food_set = {"crepe", "salad", "tsunamayo", "yakitori"}
new_food_set = {"sushi", "salad", "tonsoku"}
food_set | new_food_set
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
```

上記プログラムにおいて，記号`|` は2つの集合の和集合を求める演算です．これは単純に2つの集合の要素を結合しているだけではありません．`"salad"` は`food_set` と `new_food_set` の両方に含まれていますが，和集合では，1つの要素になっています．集合では同じ値の要素を複数個含めることができないことに注意しましょう．

<font color="OrangeRed">(っ´･ω･)っ</font><font color="IndianRed">【集合に関するその他の演算については[Python公式マニュアル set()関数](https://docs.python.org/ja/3/library/stdtypes.html#set)を参照ください．】</font>

## 2-6. <font color="Teal">繰り返し処理</font>

### 2-6-1. <font color="Teal">for文による繰り返し</font>

Pythonのfor文は，リスト内のすべての要素に対し同じ処理をするというものです．  
以下のような形式で書きます．
```
for 変数 in リスト:
    変数を使った処理
```

<font color="RoyalBlue">【実習】 食品名が格納された `foods` というリストを生成し，それを1行ずつ出力する．</font>

```
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
foods = ["salad", "hamburger", "french fries", "fried chicken"]
for food in foods:
    print(food)
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
```

以下の部分がループ処理のための記述です．
```
for food in foods:
    print(food)
```

2行めの字下げされている部分が繰り返し処理される部分です．  
繰り返し処理において，リスト`foods`の各要素が`food`という変数に代入され，変数`food` の値の出力を行っています．  
ここで，2行めの字下げは見やすさのためだけに行っているわけではなく，Pythonの文法の一部として行っています．なので，この字下げを省略することはできません．  
字下げのことを「インデント」と呼びます．

<font color="OrangeRed">(っ´･ω･)っ</font><font color="IndianRed">【インデント（字下げ）については様々なやり方ができますが，[PEP8](https://pep8-ja.readthedocs.io/ja/latest/#id6)では，半角のスペース4つで行うことが推奨されています．この教材でも，4つのスペースでインデントを行います．】</font>

### 2-6-2. <font color="Teal">while文による繰り返し</font>

指定した条件を満たしている間，繰り返し処理を行うようにする方法として，while文があります．  
while文は以下のような形式で書きます．
```
while 条件式:
    処理
```
条件式が満たされている限りインデントされたブロックに記述された処理が繰り返し実行されます．

<font color="RoyalBlue">【実習】 `1` から `5` までの数字を順に出力する処理を `while` 文を使って書く．</font>

```
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
num = 1
while num <= 5:
    print(num)
    num +=1
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
```

上記のように書くと，`num` の値を出力しさらに `1` だけ大きくする処理を `num` が `5` 以下の間継続します．

### 2-6-3. <font color="Teal">数値列によるfor文</font>

変数の値を0から10まで1ずつ増やしてループを回したいような場合，0から10の整数を要素とするようなリストを作ればfor文を使用することができます．  
関数 `range()` を利用すると連続した数値の列（リストのようなもの）を生成することができます．  
例えば，以下のように記述することで，1以上で5未満の整数の列，すなわち `1, 2, 3, 4` を生成することができます．
```
range(1, 5)
```


<font color="OrangeRed">(っ´･ω･)っ</font><font color="IndianRed">【`range(1,5)` と書いた時，数字 1 は含まれるが，数字 5 は含まれないことに注意しましょう．これはリストのスライスと同様の仕様です．】</font>

`range()` で生成した数値列を `for` 文で使用することができます．

<font color="RoyalBlue">【実習】 1から4までの数値を出力する処理を `range()` を利用して記載する．</font>

```
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
for num in range(1, 5):
    print(num)
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
```

関数 `range()` において，括弧の中に数値を一つだけ指定すると，生成される整数列の先頭の数値が 0 になります．

<font color="RoyalBlue">【実習】 0から4までの数値を出力する処理を `range()` を利用して記載する．</font>

```
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
for num in range(5):
    print(num)
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
```

<font color="OrangeRed">(っ´･ω･)っ</font><font color="IndianRed">【`range(1,5)` と書いた時，1 以上 5 未満の整数値からなるリストが生成されると思われた方もいるかもしれませんが，厳密には少し違います．実際，Python の古いバージョンである Python2 では，リストが生成されていたのですが，現行の Python3 ではリストとは少し異なるものが生成されるようになりました．`range()` で指定する値の範囲が広い時，実際にリストを生成してしまうとメモリ消費量が大きくなります．なので，Python3 の `range()` では範囲内のすべての要素からなるリストを実際に生成するようなことはせず，どの値がリストに含まれているのかを計算できるようなデータ構造が生成されています（数値列は規則正しく生成されているのでこれは可能です）．なので，Python3 では，リストの代わりに `range()` を使うことでメモリの消費量を少なくすることができます．】</font>

`range()` により生成される数値列からリストを生成するためには `list()` を利用します．

<font color="RoyalBlue">【実習】 0から4までの整数値からなるリストを生成する．</font>

```
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
numbers = list(range(5))
numbers
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
```

`range()` による数列の生成ですが，指定した範囲内の数値を一定間隔で取り出すようにすることもできます．  
リストのスライスと同様に3つめの引数に間隔を指定します．

<font color="RoyalBlue">【実習】 0から50までの5の倍数のリストを生成する．</font>

```
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
multiples_of_five = list(range(0, 51, 5))
multiples_of_five
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
```

### 2-6-4. <font color="Teal">リスト内包表記</font>

リスト内包表記と呼ばれる書き方を用いることで，複雑なリストを簡便に生成することができます．  
`[変数を使った式 for 変数 in リスト]` と書くと，リストに含まれるすべての要素について，「変数を使った式」の値がリストに追加されます．

<font color="RoyalBlue">【実習】 非負の整数の3乗により生成される値を小さい順に10個生成してリストを作る．</font>

```
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
cubes = [num ** 3 for num in range(10)]
print(cubes)
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
```

<font color="OrangeRed">(っ´･ω･)っ</font><font color="IndianRed">【内包表記とは集合の定義の際に使用される書き方で，例えば，上記リストの要素からなる集合は
$\{n^3 \mid 0 \leq n <10, n \in \mathbb{Z}\}$
と書くことができます．】</font>

<font color="OrangeRed">(っ´･ω･)っ</font><font color="IndianRed">【リスト内包表記のさらに高度な使い方については，[Python公式マニュアルの「リストの内包表記」](https://docs.python.org/ja/3/tutorial/datastructures.html#list-comprehensions)をご覧ください．】</font>

与えられたデータにより異なる処理を行えるということは，ソフトウェアにとって本質的に重要です．
そのためには，指定した条件を満たす場合と満たさない場合で異なる処理を行えるようにすることが必要です．
それは条件分岐と呼ばれており，Python では`if` 文を使用して条件分岐を行います．

## 2-7. <font color="Teal">条件分岐</font>

if文を使うことで条件により処理を分岐させることができます．

### 2-7-1. <font color="Teal">if文の基本</font>

条件分岐の記述は以下のように行います．
```
if 条件式1:
    条件式1を満たす場合の処理
elif 条件式2:
    条件式2を満たす場合の処理
else:
    条件を満たさない場合の処理
```
もし条件式1が満たされた場合，すなわち，条件式1が `True` である場合，if部の「条件式1を満たす場合の処理」が実行されます．  
条件式1が満たされない場合，すなわち，条件式1が `False` である場合，
もし条件式2が `True` であるなら，elif部の「条件式2を満たす場合の処理」が実行されます．  
どちらも満たさない場合，else部の「条件を満たさない場合の処理」が実行されます．

`elif`部，`else`部は省略可能です．  
また`elif`部は2つ以上にすることも可能です．

<font color="OrangeRed">(っ´･ω･)っ</font><font color="IndianRed">【真理値を表すために[ブール値](https://docs.python.org/ja/3/library/stdtypes.html#boolean-values)が使われますが，`True` と `False` はそれぞれ真と偽を表すためのブール値の定数です．】</font>

<font color="RoyalBlue">【実習】 "coffee", "milk", "beer", "water" の 4 つの飲み物からなるリストについて `for` 文を実行し，"beer"，"coffee"，それ以外の場合でそれぞれ異なる出力を行う．</font>

```
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
drinks = ["coffee", "milk", "beer", "water"]
for drink in drinks:
    if drink == "beer":
        print("Alcohol drink")
    elif drink == "coffee":
        print("Caffeinated drink")
    else:
        print("Non-alcohol and caffeine-free drink")
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
```

`a == b` と書くと， `a` と `b` の値が等しい時 `True` ，そうでない時 `False` になります．

    <td>a が b 以上なら True</td>
  </tr>
</table>


<font color="OrangeRed">(っ´･ω･)っ</font><font color="IndianRed">【条件式が計算されると条件式自体が `True` か `False` かの値を持つようになります．条件式の真理値判定の詳細については[Python公式ドキュメント](https://docs.python.org/ja/3/library/stdtypes.html#truth)を参照ください．】</font>

<font color="OrangeRed">(っ´･ω･)っ</font><font color="IndianRed">【条件式にはリストや数値（整数，浮動小数点数），文字列などを入れることもでき，その際，空のリストや数字の 0 ，長さ 0 の文字列などは[`False` と判定されます](https://docs.python.org/ja/3/library/stdtypes.html?highlight=%E4%B8%BB%E3%81%AA%E7%B5%84%E3%81%BF%E8%BE%BC%E3%81%BF%E3%82%AA%E3%83%96%E3%82%B8%E3%82%A7%E3%82%AF%E3%83%88%E3%81%A7%E5%81%BD%E3%81%A8%E5%88%A4%E5%AE%9A%E3%81%95%E3%82%8C%E3%82%8B%E3%82%82%E3%81%AE#truth)．】</font>

### 2-7-2. <font color="Teal">複数個の条件設定</font>

ブール演算子 `and`，`or`，`not` を利用することで複数の条件を組み合わせた複雑な条件式を作ることができます．  
以下にブール演算子をまとめておきます．
<table>
  <tr align="center">
    <th>ブール演算子</th>
    <th>名称</th>
    <th>意味</th>
  </tr>
  <tr align="center">
    <td>p and q</td>
    <td>論理積</td>
    <td>p と q が両方 True の 時 True，それ以外の時 False</td>
  </tr>
  <tr align="center">
    <td>p or q</td>
    <td>論理和</td>
    <td>p か q のひとつ以上が True の 時 True，それ以外の時 False</td>
  </tr>
  <tr align="center">
    <td>not p</td>
    <td>否定</td>
    <td>p が True なら False，p が False なら True</td>
  </tr>
</table>


<font color="OrangeRed">(っ´･ω･)っ</font><font color="IndianRed">【論理積，論理和については[短絡評価が行われます](https://docs.python.org/ja/3/library/stdtypes.html?highlight=%E7%9F%AD%E7%B5%A1%E8%A9%95%E4%BE%A1#boolean-operations-and-or-not)．つまり 左側（上記 p）の真理値のみで結果が判定できる時，右側（上記 q）については値の計算を行いません．】</font>

<font color="RoyalBlue">【実習】 変数 `age`，`drink` について，`age < 20` かつ `drink == "beer"` の場合とそれ以外の場合で異なる出力処理を行う．</font>

```
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
drink = "beer"
age = 15
if age < 20 and drink == "beer":
    print("You cannot order this drink.")
else:
    print("You can order this drink.")
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
```

### 2-7-3. <font color="Teal">繰り返し処理の中断</font>

for文やwhile文などの繰り返し処理において，指定した条件を満たした場合に処理を中断させることができます．その場合 `break` 文を使用します．

<font color="OrangeRed">(っ´･ω･)っ</font><font color="IndianRed">【繰り返し処理が2重，3重のループになっていることがありますが，その場合．`break` 文では一番内側のループのみを抜けます．】</font>

<font color="RoyalBlue">【実習】 0から5までの整数を出力する処理を`break` 文を使用して書く．</font>

```
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
num = 0
while True:
    print(num)
    num += 1
    if num > 5:
        break
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
```

<font color="OrangeRed">(っ´･ω･)っ</font><font color="IndianRed">【上記のプログラムにおいて `while` 文の条件式が `True` になっていますが，これは，常に繰り返しを継続するという意味です．つまり上記のプログラムでは，`break` 文だけが繰り返し処理を終了させることができます．】</font>

<font color="OrangeRed">(っ´･ω･)っ</font><font color="IndianRed">【条件に応じて繰り返し処理の一部をスキップする処理として [`continue`文](https://docs.python.org/ja/3/tutorial/controlflow.html?highlight=continue#break-and-continue-statements-and-else-clauses-on-loops) がありますが，ここでは省略します．また，何もしないことを明示的に示す文として [`pass`文](https://docs.python.org/ja/3/tutorial/controlflow.html?highlight=pass#pass-statements) がありますが，これも省略します．】</font>



## 2-8. <font color="Teal">関数</font>

### 2-8-1. <font color="Teal">関数の定義</font>

関数を定義するには， `def 関数名():` と書き，その下にインデントして処理を書きます．  
他の場所から関数を呼び出す際は，関数を括弧付きで記述します．

<font color="RoyalBlue">【実習】 あいさつを表示する関数 `say_hello()` を作ったのち，それを2回呼び出す．</font>

```
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
def say_hello():
    print("Hey!")
    print("How are you?")

say_hello()
say_hello()
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
```

<font color="OrangeRed">(っ´･ω･)っ</font><font color="IndianRed">【定義した関数等の説明文を docstring と呼びます．docstring の書き方に関してはいくつかの慣習があり，それに従うことでソースコードから仕様書を自動生成したりすることができるようになります．例えば docstring を書く際のコメントアウトには，たとえ 1 行であっても，慣習的に `#` より `"""` が使われます．詳しくは [docstringの説明（PEP257）](https://www.python.org/dev/peps/pep-0257/) をご覧ください．】</font>

### 2-8-2. <font color="Teal">引数の設定</font>

関数にデータを渡して関数内で使用できるようにすることができます．  
関数定義時に括弧の中に変数を書きます．

<font color="RoyalBlue">【実習】 名前とあいさつを表示する関数 `say_hello(name)` を作り， `name` に色々な値を渡してみる．</font>

```
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
def say_hello(name):
    print("Hey! {}．".format(name))
    print("How are you?")

say_hello("Alice")
person = "Bob"
say_hello(person)
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
```

関数定義時に使われている変数（上記では，`def say_hello(name）:` の　`name`）のことを Python では仮引数と呼んでいます．

一方で，呼び出し元で，関数に渡すために使われている値（`"Alice"` や `person`）を実引数と呼んでいます．

つまり，関数呼び出し時，実引数の値が仮引数に代入されます．

<font color="OrangeRed">(っ´･ω･)っ</font><font color="IndianRed">【引数は「ひきすう」と読みます．】</font>

### 2-8-3. <font color="Teal">仮引数の値変更</font>

関数内で仮引数の値を変更する場合は注意が必要です．

関数内で仮引数の値を変更しても基本的には実引数（呼び出し元の変数）の値は変化しません．
これは，仮引数に実引数の値が代入されているためです（仮引数と実引数は別々に値を管理します）．

<font color="OrangeRed">(っ´･ω･)っ</font><font color="IndianRed">【実引数にリストや辞書等が指定された場合には少し異なることが起きます（実変数のリストを編集可能になります）が，ここでは説明を省略します．】</font>


<font color="RoyalBlue">【実習】 仮引数の値を変更しても実引数の値が変わらないことを確認する．</font>

```
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
def display_triple_value(num):
    num *= 3
    print("Triple value is {}.".format(num))

score = 20
display_triple_value(score)
print("The original score is {}.".format(score))
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
```

### 2-8-4. <font color="Teal">位置引数とキーワード引数</font>

引数は複数個設定することができます．  
その際，関数呼び出し時の引数の区別の仕方として，位置によるものとキーワードによるものがあります．  
それぞれ[位置引数，キーワード引数](https://docs.python.org/ja/3/glossary.html#term-argument)と呼ばれます．

位置引数を使用する場合，関数呼び出し元の引数（実引数）の 1 つめは，関数側の引数（仮引数）の 1 つめに代入され，
2 つめの実引数は， 2 つめの仮引数に代入されます．以下同様です．

<font color="RoyalBlue">【実習】 位置引数により，複数の値を関数に渡す．</font>

```
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
def display_score(my_score, perfect_score):
    print("I scored {0} out of {1} points.".format(score, perfect_score))

display_score(75, 100)
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
```

キーワード引数を使用する場合，関数呼び出し時に括弧の中で仮引数を指定して値を代入します．

<font color="RoyalBlue">【実習】 キーワード引数により，複数の値を関数に渡す．</font>

```
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
def display_score(my_score, perfect_score):
    print("I scored {0} out of {1} points.".format(score, perfect_score))

display_score(perfect_score=100, my_score=75)
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
```

<font color="OrangeRed">(っ´･ω･)っ</font><font color="IndianRed">【関数の呼び出し時，位置引数とキーワード引数を混在させることもできますが，その場合，キーワード引数は位置引数の後ろに書く必要がありますので注意しましょう．】</font>

### 2-8-5. <font color="Teal">デフォルト値の設定</font>

仮引数にはデフォルトの値を設定することができます．

デフォルト値が設定されている仮引数については，関数呼び出し時に実引数の指定を省略できるようになります．
省略された場合は，デフォルト値が使用されます．

<font color="RoyalBlue">【実習】 デフォルト値が設定された仮引数について，実引数の指定を省略する．</font>

```
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
def display_score(my_score, perfect_score=100):
    print("I scored {0} out of {1} points.".format(my_score, perfect_score))

display_score(75)
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
```

デフォルト値から変更したい場合は，これまで通り実引数を指定します．

<font color="OrangeRed">(っ´･ω･)っ</font><font color="IndianRed">【引数についてはさらに高度な設定（例えば，引数の個数を呼び出し元で変えられるようにする機能等）が可能ですが，ここでは省略します．詳しく知りたい方はPythonチュートリアルの[関数定義についてもう少し](https://docs.python.org/ja/3/tutorial/controlflow.html#more-on-defining-functions)を参照してください．】</font>

### 2-8-6. <font color="Teal">関数の戻り値</font>

関数内で計算した値を呼び出し元に戻してあげることも可能です．  
そのような値を戻り値と呼びます．

関数内で `return 値` のように書くと，その値を呼び出し元に渡して，関数の処理を終了します．

<font color="RoyalBlue">【実習】 関数の呼び出し元に戻り値を返す．</font>

```
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
def is_positive_number(num):
    return (num > 0)

result = is_positive_number(-5)
result
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
```

### 2-8-7. <font color="Teal">関数内で辞書生成</font>

関数の中で，リストや辞書等を作成してそれを戻り値にするということもできます．

<font color="RoyalBlue">【実習】 関数の中で辞書を作り，それを戻り値とする．</font>

```
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
def get_exam_result(score):
    if score >= 80:
        grade = "A"
    elif score >= 70:
        grade = "B"
    elif score >= 60:
        grade = "C"
    else:
        grade = "D"
    result = {"score": score, "grade": grade}
    return result

alice = get_exam_result(72)
print(alice)
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
```

<font color="OrangeRed">(っ´･ω･)っ</font><font color="IndianRed">【この挙動はC言語等を使っていてスタック領域などの知識がある方にとっては少し違和感があるかもしれませんが，Python の変数は，記憶領域を確保している訳ではなく，生成されたオブジェクトのIDを覚えているだけだということが分かれば理解できると思います．】</font>

## 2-9. <font color="Teal">ファイル入出力</font>

外部からデータを取り込む場合や，プログラムによる処理結果をプログラム終了後も保持したい場合，ファイルを使用することができます．

### 2-9-1. <font color="Teal">ファイルの読み込み</font>

ここでは，以下の内容（グリム童話「Little Red-Cap」（赤ずきん；著作権フリー）の一部）が書かれたファイル little_red_cap.txt を作成したのち，Pythonプログラムからそのファイルを読み込んでみます．ファイルの作成については，ここではLinuxのechoコマンドを使って行います．

```
‘Oh! grandmother,’ she said, ‘what big ears you have!’
‘The better to hear you with, my child,’ was the reply.
‘But, grandmother, what big eyes you have!’ she said.
‘The better to see you with, my dear.’
‘But, grandmother, what large hands you have!’
‘The better to hug you with.’
‘Oh! but, grandmother, what a terrible big mouth you have!’
‘The better to eat you with!’
```

<font color="RoyalBlue">【実習】 上記内容が書かれたテキストファイル little_red_cap.txt を作成する．</font>

```
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
!echo -e  "‘Oh! grandmother,’ she said, ‘what big ears you have!’\n\
‘The better to hear you with, my child,’ was the reply.\n\
‘But, grandmother, what big eyes you have!’ she said.\n\
‘The better to see you with, my dear.’\n\
‘But, grandmother, what large hands you have!’\n\
‘The better to hug you with.’\n\
‘Oh! but, grandmother, what a terrible big mouth you have!’\n\
‘The better to eat you with!’" > little_red_cap.txt
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
```

<font color="OrangeRed">(っ´･ω･)っ</font><font color="IndianRed">【Colab（の裏で動いているIPython）では，先頭に `!` をつけることで，シェルコマンド（Linuxコマンド）を実行することができます．詳細は[IPythonの公式ドキュメント](https://ipython.readthedocs.io/en/stable/interactive/python-ipython-diff.html?highlight=exclamation#shell-assignment)をご覧ください．上記ではリダイレクトという機能を使用してファイルを作成しています．】</font>

Linuxの`cat`コマンドを使用することで，ファイルの中身を表示させることができます．

<font color="RoyalBlue">【実習】 テキストファイル little_red_cap.txt が正しく作成されていることを確認する．</font>

```
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
!cat little_red_cap.txt
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
```

上記に書いたファイルの内容が表示されていれば，正しくファイルが作成されています．

<font color="OrangeRed">(っ´･ω･)っ</font><font color="IndianRed">【Colabでも作成したテキストファイルの中身を確認することができます．画面左端のフォルダマークをクリックしたのち，所望のファイルをダブルクリックするとファイルの内容が表示されます．】</font>

次に，Pythonプログラムから作成したファイルを読み込んでみます．   
ファイルのオープンは`open()`，クローズは`close()`を使って行うことができます．  
ファイルを開いた際，処理が終わったら必ずファイルを閉じる必要がありますが，[`with`文](https://docs.python.org/ja/3/reference/compound_stmts.html#the-with-statement)を使うことで，`close()`を呼ばなくてもwithブロック終了後に必ずファイルを閉じるようにすることができます．


<font color="RoyalBlue">【実習】 テキストファイル little_red_cap.txt をPythonで読み込み，1行ずつ出力する．</font>

```
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
with open("little_red_cap.txt", "r") as f:
    for line in f:
        print(line)
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
```

上のコードにおいて，`open("little_red_cap.txt", "r")` は 読み込みモード（`r`）でファイル `little_red_cap.txt`を開くという記述です．  
ファイルのオープンに成功するとファイルオブジェクトというものが生成され，それが変数 `f` に代入されます．  
Python プログラムからこのファイルにアクセスするためには，変数`f` を利用します．


変数 `f` に対して以下のように `for` 文を用いると，ファイルを先頭から順に1行ずつ読み込むことができます．
```
for line in f:
    各行に対する処理
```

<font color="OrangeRed">(っ´･ω･)っ</font><font color="IndianRed">【1行ずつファイルを読み込む際，行末の改行コードも読み込まれていることに注意してください．`print`文では出力後に自動で改行が挿入されるため，上記では，1行出力するごとに2個の改行が挿入されています．】</font>

### 2-9-2. <font color="Teal">ファイルへの新規出力</font>

ファイルへ何かを出力する際には，`open()` のオプションに書き込みモード（`w`）を指定します．  
このモードを指定すると，既にファイルが存在する場合には，元の内容が消され，存在しない場合には新たにファイルを生成します．

<font color="RoyalBlue">【実習】 テキストファイル programming.txt を書き込みモードで開き，文を2行書き込む</font>

```
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
with open("programming2.txt", "w") as f:
    f.write("I study computer science.\n")
    f.write("I like programming.\n")
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
```

ファイルに書き込みを行うには，ファイルオブジェクト `f` に対して，`write()` メソッドを使います．

<font color="OrangeRed">(っ´･ω･)っ</font><font color="IndianRed">【改行したい場合には，明示的に改行記号`\n` を書く必要がありますので注意しましょう．】</font>

<font color="RoyalBlue">【実習】 作成したファイル `programming.txt` の中身を確認する．</font>

```
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
!cat programming.txt
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
```

2行に渡って文が書き込まれていれば成功です．

### 2-9-3. <font color="Teal">ファイルへの追加出力</font>

既存のファイルの末尾に内容を追記したい場合には，追記モード `a` を指定します．

<font color="RoyalBlue">【実習】 ファイル `programming.txt` に追記を行う．</font>

## 2-10. <font color="Teal">クラス</font>

ソフトウェア設計手法のひとつにオブジェクト指向があります．  
オブジェクト指向では，ソフトウェアをオブジェクトに分割し，オブジェクトごとに役割分担を決めます．  
それにより，可読性，保守性の高いソフトウェアを開発できるようにします．

クラスとはオブジェクトの設計図にあたるものです．  
クラスでは，オブジェクトが管理するデータとそのデータに対する操作（提供される機能）を指定します．

そして，その設計図に従って実体化された変数をインスタンスと呼びます．

### 2-10-1. <font color="Teal">クラスの生成</font>

シンプルな例として，犬の名前と年齢を管理するクラス `Dog` を作ってみます．

クラス `Dog` は以下のように作成します．

```
class Dog:
    def __init__(self, name, age):
        self.name = name
        self.age  = age
```
詳しいことはこの後で説明しますが，このように書くことで，インスタンス内で，`name`と`age`の 2 つの変数が管理されるようになります．

また，作成したクラス `Dog` のインスタンスを生成するには以下のように書きます．
```
mydog = Dog("Ken", 6)
```

ここでは，クラス `Dog` のインスタンス `mydog` を生成し，名前 `name` を `"ken"`，年齢 `age` を`6` に設定しています．これも詳しいことは後で説明しますが，`Dog` の後に指定している引数は，`__init__()`に書かれた引数のうち，先頭の `self` を除くもの（つまり，`name` と `age`）に対応しています．

インスタンス `mydog` が管理している変数 `name` にアクセスするには， `mydog.name` と書きます．



<font color="RoyalBlue">【実習】 犬の名前と年齢を保持するクラス `Dog` を作成後，そのインスタンス `mydog`を生成し，名前を `"Ken"`，年齢を `6` とする．</font>

```
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
class Dog:
    def __init__(self, name, age):
        self.name = name
        self.age  = age

mydog = Dog("Ken", 6)
print(mydog.name)
print(mydog.age)
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
```

<font color="OrangeRed">(っ´･ω･)っ</font><font color="IndianRed">【クラスの名前について，[PEP8](https://pep8-ja.readthedocs.io/ja/latest/#id32)では，先頭文字を大文字にすることが推奨されています．この教材でもそれに従います．クラスの名前を大文字始まりにし，インスタンスの名前を小文字始まりにすることで，ソースコード上でもクラスとインスタンスの違いが分かりやすくなります．】</font>

それでは，クラス作成処理の詳細をみていきます．上記のソースコードの一部を再掲しておきます．

```
class Dog:
    def __init__(self, name, age):
        self.name = name
        self.age  = age

mydog = Dog("Ken", 6)
```
最初の行には，作成するクラス名を書き，次の行からインデントをしてクラスの中身を書きます．
このクラスでは，クラスの中身として `__init__()` という名前の関数が定義されています．
Python では，クラス内で定義される関数のことをメソッドと呼んでいますので，これからは，それに従うことにします．つまり，`Dog` クラスの中に`__init__()` メソッドが定義されているということになります．

`__init__()` メソッドには，インスタンス生成時に行う処理を書きます．ここでは`self.name` に `name`の値を代入していますが，`self` については特殊な仕様になっていますので，それについて説明します．

クラス内で定義されるすべてのメソッドについて，先頭の引数は特別な意味を持ちます．[先頭の引数は`self`と呼ばれており](https://docs.python.org/ja/3/tutorial/classes.html#random-remarks)，`self` は，オブジェクト自身（インスタンス自身）を表します．
よって上記の `self.name = name` という処理はインスタンスで管理する `name` という変数に引数で指定された変数 `name` を代入するという意味になります．

`self` に関する注意事項ですが，メソッドの呼び出し元では，この引数を指定する必要はありません．
よって，メソッドを呼び出す際，メソッドの定義時と比較し，引数の数が 1 つ少なくなります．



<font color="OrangeRed">(っ´･ω･)っ</font><font color="IndianRed">【インスタンス自身を表すワードとして，C++やC#では `this`，Rubyでは `self` が使用されています．Pythonに関しては，文法として `self` というワードが決められているわけではなく，メソッドの先頭引数でユーザが名前を指定することになっています．そのため `self`以外のワードを使用することもできますが，[`self` という名前を使う慣習になっています](https://docs.python.org/ja/3/tutorial/classes.html#random-remarks)ので，他の名前を使うと，他の人が読みにくいコードになってしまうかもしれません．[PEP8](https://pep8-ja.readthedocs.io/ja/latest/#id37)でも `self` を使うように定められています．】</font>

`__init__()` メソッドについて，もう少し詳しく述べておきます．
クラスにおいて `__init__()` メソッドは特別な意味を持っています．
`__init__()` メソッドが定義されている場合，インスタンス生成時に，自動的に`__init__()` メソッドが呼び出されます．オブジェクト指向プログラミングでは，このようなメソッドをコンストラクタと呼びます．

インスタンスの生成は，クラス名の後ろに括弧をつけて，その中に引数を指定することで行います．
上記のプログラムで `mydog = Dog("Ken", 6)` という処理がありますが，`Dog("Ken", 6)` は，インスタンス生成後，この 2 つの引数を用いて`__init__()` メソッドを呼び出すという意味になります．
`__init__()` メソッドの定義には引数が 3 つありますが，呼び出し元では，`self`を指定する必要がないため引数は 2 つになります．

<font color="OrangeRed">(っ´･ω･)っ</font><font color="IndianRed">【以上が，クラスの基本的な使い方の説明になりますが，複数のデータ（ここでは，`name` と ` age`）をまとめて管理できることがクラスを使うメリットの一つです．クラスを使わない場合，個々の変数を個別に管理することになり，プログラムが煩雑になります．今回の例では，クラスを使うことでかえって処理が複雑になっていると感じるかもしれませんが，プログラムが大規模化すればするほど，クラスを作ることのメリットを強く感じられるようになると思います．】</font>

### 2-10-2. <font color="Teal">メソッド</font>

クラス内でメソッドを定義すると，それはクラスで管理するメソッドになります．それを呼び出すには， `インスタンス名.メソッド名()` のように書きます．

<font color="RoyalBlue">【実習】 犬の名前と年齢を保持するクラス `Dog` に`sit()` というメソッドを追加する．</font>

```
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
class Dog:
    def __init__(self, name, age):
        self.name = name
        self.age  = age
    def sit(self):
        print("{0} is now sitting.".format(self.name))

mydog = Dog("Ken", 6)
mydog.sit()
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
```

<font color="OrangeRed">(っ´･ω･)っ</font><font color="IndianRed">【クラス内でメソッドを定義する際は，必ず先頭の引数を `self` にしないといけないことに注意しましょう．メソッド呼び出し時にはこの引数を指定する必要がないため，`mydog.sit()` のように引数が 0 個になっています．】</font>

上記では `mydog` という1つのインスタンスで，複数個の変数を管理でき，これらの変数に関連するメソッドも使用できます．これはソースコードの可読性を向上させるために役立ちます．

以下のようにインスタンスを何個も生成するような処理ではさらに有用です．

<font color="RoyalBlue">【実習】 犬の名前と年齢を保持するクラス `Dog` のインスタンスを複数生成する．</font>

```
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
class Dog:
    def __init__(self, name, age):
        self.name = name
        self.age  = age
    def sit(self):
        print("{0} is now sitting.".format(self.name))

mydog = Dog("Ken", 6)
yourdog = Dog("Peko", 4)
mydog.sit()
yourdog.sit()
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
```

### 2-10-3. <font color="Teal">クラス変数</font>

`インスタンス名. 変数名` もしくは `self.変数名` とすると，インスタンスが保持する変数にアクセスすることができましたが，そうではなく，クラスから生成されたすべてのインスタンスで値が共有されるような変数を作ることができます．そのような変数をクラス変数と呼びます．

逆に，これまで説明してきたような，インスタンスごとに管理される変数はインスタンス変数と呼びます．

クラス変数を作成するには，単にクラスの定義部で変数に値を代入します．クラス変数にアクセスするには， `クラス名.変数名` とします．別のアクセス方法として，インスタンス変数に同じ名前の変数がない場合には，`インスタンス名.変数名` と書いてもクラス変数にアクセスすることができます（ただし同名のインスタンス変数がある場合には，そちらにアクセスしますので注意が必要です）．

<font color="RoyalBlue">【実習】 犬情報を保持するクラス `Dog` にクラス変数 `type` を追加する．</font>

```
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
class Dog:
    type = "dog"
    def __init__(self, name, age):
        self.name = name
        self.age  = age
    def sit(self):
        print("{0} is now sitting.".format(self.name))

mydog = Dog("Ken", 6)
yourdog = Dog("Peko", 4)
print(Dog.type)
print(yourdog.type)
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
```

クラス変数には[きちんと仕組みを把握していないと理解し難いような挙動がある](https://docs.python.org/ja/3/tutorial/classes.html#class-and-instance-variables)ので注意しましょう．
クラス変数の代わりにインスタンス変数を使うことができる場合は，インスタンス変数を使った方が無難だと思われます．

<font color="OrangeRed">(っ´･ω･)っ</font><font color="IndianRed">【インスタンス変数を作るつもりが，書き方を間違えてクラス変数ができてしまうことがあるので，注意しましょう．インスタンス変数を作るには`__init__()` メソッドの中で `self.インスタンス名 = 値` のように記載します（どのメソッドの中でもインスタンス変数を作成できますが，インスタンス初期化の処理で作成するのが分かりやすいと思います）．クラス定義の中で `self`　を付けずに変数に値を代入すると，その変数はクラス変数になってしまいます．C++言語などでクラスを実装している方は特に注意が必要です．C++言語と同様の書き方でインスタンス変数を作成しようとすると，クラス変数が作成されてしまいます． また，`インスタンス名.変数名` という書き方でもクラス変数にアクセスできてしまうため，間違っていても実行時にエラーが発生せず，間違いに気づきにくくなっています．】</font>

### 2-10-4. <font color="Teal">プライベート変数</font>

オブジェクト指向プログラミングでは，インスタンスへの外部からのアクセスを適切に制限することで，モジュールごとの権限と役割分担を明確にすることができます．例えば，ある外部モジュールがインスタンス内のすべてのデータにアクセス可能だとすると，本来インスタンス内で行うべき処理まで，その外部モジュールがやってしまう恐れがあり，役割分担が崩壊する可能性があります．データにアクセス制限がかかっている場合，そのデータを使う処理は，そのデータにアクセス可能なモジュールで行うしかしかなくなり，適切な役割分担が維持されます．

クラスにおいて，外部からのアクセスが禁止されている変数をプライベート変数と呼びます．また，外部からのアクセスが禁止されているメソッドをプライベートメソッドと呼びます．

Pythonでは，プライベート変数やプライベートメソッドを作ることはできません．つまり，どの変数やメソッドも外部からアクセス可能です．
ただし，[変数やメソッドの名前を変えることで，プライベートとして扱われるべき変数やメソッドを区別するという慣習があります](https://docs.python.org/ja/3/tutorial/classes.html#private-variables)．実装を行う人がプライベートとして扱われるべき変数等に外部からアクセスしないようにすることで，本来の目的を達成することができます．

Pythonの慣習では，プライベートとして扱われるべき変数やメソッドの先頭文字をアンダースコア（_）にします．

<font color="RoyalBlue">【実習】 試験の点数を保持し試験に合格したかどうかを答えるクラスを作成する．ただし，スコア自体には外部からアクセスしないことを推奨するような書き方にする．</font>

```
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
class PrivateScore:
    def __init__(self, score):
        self._score = score
    def is_passed(self):
        if self._score >= 60:
            return True
        else:
            return False

score_math = PrivateScore(55)
print(score_math.is_passed())
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
```

<font color="OrangeRed">(っ´･ω･)っ</font><font color="IndianRed">【上記プログラムでは，インスタンス変数`_score` をプライベートの扱いにしています（変数名の先頭にアンダースコアを付けています）．プライベート変数に外部からアクセスしないようにすることは，保守性を高めるのにも役立ちます．上記プログラムにおいて，合格の基準を60点から70点に変更する場合， `is_passed()` 中の60を70に変更するだけで，修正が完了しますが，もし，スコア自体を外部から取得できるようになっていた場合には，外部で勝手に合否判定処理を書いているかもしれないため，スコアを取得しているすべてのモジュールについて修正の必要性をチェックしなければならなくなります．】</font>

### 2-10-5. <font color="Teal">継承</font>

あるクラスのインスタンスを複数生成する際，一部のインスタンスのみ処理を変更したり，別の処理を追加したりしたくなることがあります．その場合，そのインスタンス用に新しくクラスを作ることもできますが，元のクラスと大部分が同じである場合，同じコードが大量に複製されることになり，コードが無駄に長くなります．同じ処理を行うコードが複数箇所に存在する場合，不具合を修正する場合にもすべての箇所を修正しないと行けなくなるため，正しく修正することが難しくなります．

このような場合の別の方法として，元のクラスからの変更部分のみを記述するというやり方があります．
このようなやり方を継承とよんでおり，修正のベースとなる元のクラスを基底クラス（もしくは親クラス），新しく作られるクラスを派生クラス（もしくは子クラス）と呼びます．

<font color="OrangeRed">(っ´･ω･)っ</font><font color="IndianRed">【基底クラスとして，複数のクラスを指定することもできます．その場合の継承を多重継承と呼びます．Python では多重継承をサポートしています．詳細についてはPython公式ドキュメントの[9.5.1. 多重継承](https://docs.python.org/ja/3/tutorial/classes.html#multiple-inheritance)をご覧ください．】</font>

Python において，継承を使ってクラス定義を行う場合，作成するクラスの名前の後ろに括弧を書き，その中に基底クラスの名前を書きます．
そして，クラス定義の中には，新規に追加するメソッドと，基底クラスから変更を行うメソッドについてのみ，メソッドの定義を書きます．

まずは，派生クラスに対して，新規にメソッドを追加してみます．

また，クラスの数が増えてきましたので，クラスに説明文（docstring）を追加してみましょう．

<font color="RoyalBlue">【実習】 犬情報を保持するクラス `Dog` を作成し，さらにそれを継承して，芸の訓練をされた犬のクラス `TrainedDog` を作成する．クラス `TrainedDog` には，年齢の数だけ"bow"と吠える処理 `do_tricks()` を追加する．</font>

```
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
class Dog:
    """ Base class for dogs """
    def __init__(self, name, age):
        self.name = name
        self.age  = age
    def sit(self):
        print("{0} is now sitting.".format(self.name))

class TrainedDog(Dog):
    """ Derived class from Dog """
    def do_tricks(self):
        for i in range(self.age):
            print("bow")

mydog = Dog("Ken", 6)
yourdog = TrainedDog("Peko", 4)
mydog.sit()
yourdog.sit()
yourdog.do_tricks()
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
```

<font color="OrangeRed">(っ´･ω･)っ</font><font color="IndianRed">【説明文（docstring）のためのコメントアウトは 1 行であっても，`#`でなく，`"""` を使う慣習でしたね[（PEP257）](https://www.python.org/dev/peps/pep-0257/) ．】</font>

クラス `TrainedDog` の定義には，`do_tricks()` メソッドしか書かれていませんが，クラス `Dog` を継承しているため， `sit()`メソッドが使えていることに注目してください．

### 2-10-6. <font color="Teal">オーバーライド</font>

オブジェクト指向プログラミングでは，基底クラスの処理を派生クラスで書き換えることをオーバーライドと呼びます．

オーバーライドの例を見てみます．

<font color="RoyalBlue">【実習】 犬情報を保持するクラス `Dog` を作成し，さらにそれを継承して，芸の訓練をされた犬のクラス `TrainedDog` を作成する．クラス `TrainedDog` では，メソッド`sit()` の処理を変えて「お手」をするようにする．</font>

```
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
class Dog:
    """ Base class for dogs """
    def __init__(self, name, age):
        self.name = name
        self.age  = age
    def sit(self):
        print("{0} is now sitting.".format(self.name))

class TrainedDog(Dog):
    """ Derived class from Dog """
    def sit(self):
        print("{0} is now sitting and giving its paw.".format(self.name))

mydog = Dog("Ken", 6)
yourdog = TrainedDog("Peko", 4)
mydog.sit()
yourdog.sit()
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
```

メソッド `sit()` について，派生クラスのインスタンスでは，処理が変わっていることに注目してください．

次に派生クラスにおいて，インスタンス変数を増やす例をみてみます．この場合，コンストラクタ `__init__()` をオーバーライドする必要があります．
この時，基底クラスのメソッドを呼べると便利です．基底クラスのメソッドを呼ぶには `super().メソッド名` のように書きます．

<font color="RoyalBlue">【実習】 犬情報を保持するクラス `Dog` を作成し，さらにそれを継承して，芸の訓練をされた犬のクラス `TrainedDog` を作成する．クラス `TrainedDog` では，インスタンス変数に飼い主の名前を保持することにし，インスタンス生成時にそれを渡すことにする．また，メソッド`sit()` の処理を変えて飼い主に「お手」をするようにする．</font>

```
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
class Dog:
    """ Base class for dogs """
    def __init__(self, name, age):
        self.name = name
        self.age  = age
    def sit(self):
        print("{0} is now sitting.".format(self.name))

class TrainedDog(Dog):
    """ Derived class from Dog """
    def __init__(self, name, age, trainer):
        super().__init__(name, age)
        self.trainer = trainer
    def sit(self):
        print("{0} is now sitting and giving its paw to {1}.".format(self.name, self.trainer))

mydog = Dog("Ken", 6)
yourdog = TrainedDog("Peko", 4, "Alice")
mydog.sit()
yourdog.sit()
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
```

<font color="OrangeRed">(っ´･ω･)っ</font><font color="IndianRed">【`super()` において引数に何の値も指定されていないことを不思議に思った方がいるかもしれません．`super()` は基底クラスを求める関数ですが，引数に何も指定しないと，どの派生クラスの基底クラスを求めればいいのかが分かりません．実はPython3 から特殊な仕様が組み込まれており，`super()` を引数なしで呼んだ場合，`super(__class__, <firstarg>)`と解釈されます．ここで，`__class__` は呼び出し元のメソッドが所属しているクラス，`<firstarg>` は呼び出し元メソッドの先頭引数（名前はユーザーにより決定されますが，一般的には`self`）を表します．つまり，`super()` を引数なしで呼んだ場合でも，実際には，クラスとインスタンスが関数に渡されています．上記のコードでは，`super()` と書く代わりに `super(__class__, self)` と書いても同じ挙動になります．こちらの方が分かりやすいかもしれませんが，コードが長くなり読みにくくなるかもしれません．詳細については．[PEP3135](https://www.python.org/dev/peps/pep-3135/)をご覧ください．】</font>

Pythonでは文字列に対して大文字小文字を変換する様々なメソッドがあります．

<font color="OrangeRed">(っ´･ω･)っ</font><font color="IndianRed">【文字列等のオブジェクトに対して実行可能な処理のことを Python ではメソッド（method）と呼びます．】</font>

### 3-3-2. <font color="Teal">空白文字の除去</font>

</font>

```
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
language = "     Python     "
print("Enjoy " + language.strip() + ".")
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
```

<font color="OrangeRed">(っ´･ω･)っ</font><font color="IndianRed">【`strip()` メソッドではスペース以外の余計な文字も取り除くことができます．詳細はPython公式ドキュメントの[str.strip()](https://docs.python.org/ja/3/library/stdtypes.html?highlight=strip#str.strip)をご覧ください．】</font>

文字列の先頭のスペースのみを取り除く場合は [lstrip()](https://docs.python.org/ja/3/library/stdtypes.html?highlight=lstrip#str.lstrip)，文字列の末尾のスペースのみを取り除く場合は [rstrip()](https://docs.python.org/ja/3/library/stdtypes.html?highlight=rstrip#str.rstrip) を使うことができます．メソッド名先頭の「r」と「l」はそれぞれ「right」と「left」を表していると思われます．

<font color="RoyalBlue">【実習】 上記の変数 `language` について，以下の4つの文字列を括弧`[  ]` で囲った文字列を出力する<br />

*   `language`
*   `language.strip()`
*   `language.lstrip()`
*   `language.rstrip()`
</font>

```
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
language = "     Python     "
print("[" + language + "]")
print("[" + language.strip() + "]")
print("[" + language.lstrip() + "]")
print("[" + language.rstrip() + "]")
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
```

<font color="OrangeRed">(っ´･ω･)っ</font><font color="IndianRed">【文字列に対しては他にも様々なメソッドが用意されています．詳細はPython公式ドキュメントの[文字列メソッド](https://docs.python.org/ja/3/library/stdtypes.html#string-methods)をご覧ください．】</font>

# 4. ✏️ <font color="Teal">正規表現</font>

文字列の中から所望のパターンを探す場合の文字列パターン指定方法のひとつに正規表現があります．  
例えば，文字列の中から2020から2029までのいずれかの数字を見つけたい場合，正規表現でパターンを `202[0-9]` と指定することで検索することができます．  
（`[0-9]`は0-9のどれかであることを示しています）

Pythonで正規表現を扱うための詳細については，[Pythonドキュメントの正規表現HOWTO](https://docs.python.org/ja/3/howto/regex.html#regex-howto)をご覧ください．

ここでは基本的なものについてのみ説明します．

## 4-1. <font color="Teal">基本的な検索方法</font>

### 4-1-1. <font color="Teal">指定パターンにマッチする文字列の取得</font>

正規表現を用いた処理を行うには，`re`モジュールをインポートします．

そして，指定した文字列パターンで文書全体を検索し，パターンにマッチする文字列をすべて取得するには，[`re.findall()`](https://docs.python.org/ja/3/library/re.html#re.findall) 関数を使用します．  
`re.findall()`では，マッチした文字列のリストが返されます（ひとつもマッチしない時は空リスト）．

<font color="RoyalBlue">【実習】 以下の文に含まれる2020から2029までの数字をすべて取得する．

`Summer Olympic Games: Brisbane 2032, Los Angeles 2028, Paris 2024`
</font>

```
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
import re

text = "Summer Olympic Games: Brisbane 2032, Los Angeles 2028, Paris 2024"
re.findall("202[0-9]", text)
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
```

### 4-1-2. <font color="Teal">最初にマッチした文字列の位置を取得</font>

パターンにマッチした文字列の文書中での位置を知りたい場合，[re.search()](https://docs.python.org/ja/3/library/re.html#re.search) 関数を使うことができます．

`re.search()` 関数では，**最初に**マッチした文字列の詳細情報を取得することができます．  
この関数からマッチオブジェクト`m`が返ってきた際，以下のように詳細情報を取り出せます．
* `m.group()`：マッチした文字列
* `m.start()`：マッチした文字列の文書中での開始位置（インデックス）
* `m.end()`：マッチした文字列の文書中での終了位置（終了文字のインデックス+1）

<font color="OrangeRed">(っ´･ω･)っ</font><font color="IndianRed">【検索を行った文書`text`に対して `text[m.start():m.end()]` と書いても，マッチした文字列（つまり`m.group()`）を取り出すことができます．】</font>

<font color="RoyalBlue">【実習】 以下の文に対して2020から2029までの数字を検索した際，最初に見つかった数字とその開始位置，終了位置を出力する．

`Summer Olympic Games: Brisbane 2032, Los Angeles 2028, Paris 2024`
</font>

```
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
m = re.search("202[0-9]", text)
print(m.group(), m.start(), m.end())
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
```

<font color="OrangeRed">(っ´･ω･)っ</font><font color="IndianRed">【[後述のグルーピング機能](#scrollTo=hAOrl0uC1cwt)を使うとマッチした文字列の一部のみを取り出せるようになります．】</font>

### 4-1-3. <font color="Teal">マッチしたすべての文字列の位置を取得</font>

`re.search()`では最初にマッチした文字列の詳細情報を知ることができましたが，マッチしたすべての文字列の詳細情報を取得するには，[`re.finditer()`](https://docs.python.org/ja/3/library/re.html#re.finditer) 関数を用います．

<font color="RoyalBlue">【実習】 以下の文に対して2020から2029までの数字を検索し，見つかったすべての数字とその開始位置，終了位置を出力する．

`Summer Olympic Games: Brisbane 2032, Los Angeles 2028, Paris 2024`
</font>

```
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
for m in re.finditer("202[0-9]", text):
    print(m.group(), m.start(), m.end())
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
```

## 4-2. <font color="Teal">正規表現における特殊文字</font>

様々な文字列パターンで検索できるようにするため，特殊文字が定められています．  
特殊文字は通常の文字としては解釈されず，特別な検索パターンを作成するために使用されます．

特殊文字の一覧を下記に示します．

    . ^ $ * + ? { } [ ] \ | ( )

なお，特殊文字を通常文字として使いたい場合（例えば特殊文字としての`.`でなく，ピリオドとしての`.`を使いたい場合），特殊文字の前にバックスラッシュを書きます（`\.`と書くことで通常文字としてのピリオドを表すことができます）．

### 4-2-1. <font color="Teal">複数文字へのマッチング</font>

#### <font color="Teal">任意文字</font>

`.` は（改行文字以外の）任意の一文字を表します．例えば，検索パターンを`P.P`とすると，`P2P`や`PPP`，`P@P`などPで始まりPで終わる任意の3文字にヒットするようになります．

<font color="OrangeRed">(っ´･ω･)っ</font><font color="IndianRed">【デフォルト設定では `.` は改行文字にはマッチしませんが，検索時の引数に [flags=re.DOTALL](https://docs.python.org/ja/3/library/re.html#re.DOTALL) を設定することで改行文字にもマッチするようになります．】</font>

<font color="RoyalBlue">【実習】 以下の文字列に対して，パターン `"P.P"` にマッチする部分をすべて取得する．

`P2P, PPP, P@P, PP, PEAP`
</font>

```
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
re.findall("P.P", "P2P, PPP, P@P, PP, PEAP")
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
```

#### <font color="Teal">指定した複数文字</font>

任意の文字でなく，指定した複数文字のいずれかにマッチさせるようにするには，`[  ]`の中に文字を指定します．検索パターンを`r[iau]ng`とすると，`ring`，`rang`，`rung`のいずれかにマッチするようになります．  
この時，連続する数字やアルファベットについては，`[0-9]`，`[A-D]`，`[x-z]` のように略して書くことができます．

また，`[^     ]`のように括弧の中の最初の文字を`^`にすると，マッチさせたくない文字を指定することができます．例えば，検索パターンを`r[^ua]n`とすると，`rnn`，`rdn`にはマッチしますが，`run`や`ran`にはマッチしません．

<font color="RoyalBlue">【実習】 以下の文字列に対して，パターン `"r[iau]ng"` にマッチする部分をすべて取得する．

`rang, ring, rung, reng, rong`
</font>

```
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
re.findall("r[iau]ng", "rang, ring, rung, reng, rong")
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
```

#### <font color="Teal">特別な文字集合</font>

いくつかの文字集合は特別な書き方で指定することができます．  
例えば任意の数字一文字を指定するには`[0-9]`と書く代わりに`\d`と書くことができます．  
代表的なものを以下にまとめます．  

ユニコードへの対応が進められており，日本語等にも（ある程度）対応しています．  
しかし以下では，半角の英数字記号（ASCII文字コードで表せるもの）を扱うこととし，日本語の扱いについては別途最後にまとめます．

|特殊文字|説明|
|:---:|:---|
|`\d`|任意の数字1文字．ASCII文字コードなら`[0-9]`と同じ．|
|`\D`|任意の数字以外の1文字．ASCII文字コードなら`[^0-9]`と同じ．|
|`\w`|単語の一部になりうる文字1文字．ASCII文字コードなら任意の英数字（＋アンダーバー）1文字（`[a-zA-Z0-9_]`と同じ）．|
|`\W`|単語の一部にならない文字1文字．ASCII文字コードなら`[^a-zA-Z0-9_]`と同じ．|
|`\s`|任意の空白文字（スペースやタブなど）1文字．ASCII文字コードなら`[ \t\n\r\f\v]`と同じ．|
|`\S`|任意の非空白文字1文字．ASCII文字コードなら`[^ \t\n\r\f\v]`と同じ．|



Pythonでこれらの特殊文字を使う場合は注意が必要です．  
バックスラッシュ（`\`）はPythonでは特別な用途（エスケープ処理）に使用されます．  
`"\\d"`のようにバックスラッシュを二重にすることで通常文字としてのバックスラッシュを書くことができますが，別の方法として，`r"\d"`のように文字列の前に`r`を書くと，文字列中のバックスラッシュを通常文字として解釈するようになります．  
これをraw文字列記法と言います．

<font color="RoyalBlue">【実習】 以下の文に対して2020から2029までの数字を検索する．（上記の特殊文字を使ってみる）

`Summer Olympic Games: Brisbane 2032, Los Angeles 2028, Paris 2024`
</font>

```
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
re.findall(r"202\d", text)
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
```

<font color="OrangeRed">(っ´･ω･)っ</font><font color="IndianRed">【Pythonのバックスラッシュ（`\`）ですが，後続の文字によっては特別な文字（エスケープ文字）とみなされず，通常文字として解釈されます．実は `"\d"` と書いた場合も，エスケープ処理は行われず，通常のバックスラッシュ文字として解釈されています．なので，上記の場合は実はraw文字列記法を使う必要はありません．しかし，正規表現で特殊文字を使う時は，常にraw文字列記法を使うようにした方が安全です．】</font>

### 4-2-2. <font color="Teal">先頭または末尾へのマッチング</font>

文字列の先頭パターンを指定して検索する場合，指定パターンの先頭に`^`と書きます．  
例えば，検索パターン `^python` は文字列`python basic`にはヒットしますが，文字列 `basic python` にはヒットしません．

同様に，末尾のパターンを指定して検索する場合，指定パターンの末尾に`$` と書きます．  
検索パターン`python$`は文字列`basic python`にはヒットしますが，文字列`python basic`にはヒットしません．

<font color="RoyalBlue">【実習】 `"python basic"`，`"basic python"`というふたつの文字列のそれぞれに対して，パターン `r"^python"` で検索を行ってみる．
</font>

```
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
print(re.findall(r"^python", "python basic"))
print(re.findall(r"^python", "basic python"))
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
```

### 4-2-3. <font color="Teal">繰り返し</font>

特定の文字（もしくはパターン）を繰り返すような文字列を検索することもできます．  

#### <font color="Teal">繰り返し回数の条件設定</font>

`+`は直前の文字を1回以上繰り返すことを示します．  
例えば，`8+`は`8`や`88`や`888888888`などにマッチします．  

<font color="OrangeRed">(っ´･ω･)っ</font><font color="IndianRed">【文字列`888888888`をパターン`8+`で検索した場合，文字列中の`8`にも`88`にもマッチしますが，[最も長いもの（`888888888`）がマッチングの結果となります](https://docs.python.org/ja/3/howto/regex.html?highlight=正規表現を繰り返したいとき、マッチングエンジンは可能な限り何度も繰り返そうと試みます#repeating-things)．】</font>

`*`は直前の文字を1回以上繰り返すか，もしくは1回も使わないことを示します．  
例えば，`be*p`は`bp`，`bep`，`beep`，`beeeeeep`などにマッチします．

`?`は直前の文字を1回使うか，1回も使わないかのどちらかであることを示します．
例えば，`colou?r`は`color`，`colour`のいずれかの文字列にマッチします．

繰り返しの回数を範囲で指定することもできます．
`kabo{2,4}m`と書くと，`o`の繰り返し回数が2回以上4回以下になります．
すなわち，`kaboom`，`kabooom`，`kaboooom` のいずれかの文字列にマッチします．

<font color="RoyalBlue">【実習】 以下の文字列に対して，パターン `r"be*p"` にマッチする部分をすべて取得する．

`bp, bep, beep, beeeeep, bop`
</font>

```
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
re.findall(r"be*p", "bp, bep, beep, beeeeep, bop")
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
```

#### <font color="Teal">特殊文字の繰り返し</font>

特殊文字も繰り返しの指定ができます．  
例えば `.*`と書くと，任意の長さの任意の文字列（文字長0を含む）になります．  
`He.*art`は`He is a lover of art`，`Headline article`，`Heart`などの文字列にヒットします．

<font color="RoyalBlue">【実習】 `"He is a lover of art"`，`"Headline article"`，`"Heart"`という3つの文字列のそれぞれに対して，パターン `r"He.*art"` で検索を行ってみる．
</font>

```
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
print(re.findall(r"He.*art", "He is a lover of art"))
print(re.findall(r"He.*art", "Headline article"))
print(re.findall(r"He.*art", "Heart"))
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
```

### 4-2-4. <font color="Teal">グルーピング</font>

#### <font color="Teal">2文字以上のパターンの繰り返し</font>

2文字以上からなるパターンの繰り返しを記述するには，そのパターンを括弧 `(   )` で囲います．  
例えば，`a(ha)+!` と書くと，`aha!`，`ahaha!`，`ahahahaha!` などにマッチします．

`re.findall()` を用いて検索を行う際，検索パターンに括弧によるグルーピングが含まれていると，括弧で囲んだ部分のパターンのみが出力されます．  
例えば，文字列`ahaha!`をパターン`a(ha)+!`で検索（`re.findall()`）すると，`["ha"]`が出力されます．

括弧で囲んだ部分だけでなく，マッチした文字列全体を得るには，例えば，`re.finditer()`を使います．

<font color="RoyalBlue">【実習】 `"ahaha!"`という文字列に対して，パターン `r"a(ha)+!"` で検索を行い，マッチしたパターン全体を出力する．
</font>

```
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
for m in re.finditer(r"a(ha)+!", "ahaha!"):
    print(m.group())
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
```

#### <font color="Teal">2種類以上のパターンへのマッチング</font>

グルーピングの別の使い方として，`|`記号を使って`(cat|dog)`のように書くと，`cat`と`dog`の両方にマッチするようになります．  
別の例として，`cent(er|re)`と書くと，`center`と`centre`のいずれかの文字列にマッチします．

<font color="RoyalBlue">【実習】 `"center, centre."`という文字列に対して，パターン `r"cent(er|re)"` で検索を行い，マッチしたパターン全体を出力する．
</font>

```
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
for m in re.finditer(r"cent(er|re)", "center, centre."):
    print(m.group())
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
```

#### <font color="Teal">マッチした文字列への通し番号付け</font>

括弧でグルーピングをして検索を行うと，マッチした文字列の括弧内の文字列に対して通し番号がつけられます．  
前方の括弧から順に1,2,・・・という風に番号が付けられますが，`\1`，`\2`，・・・と書くことでそれを使用できます．

例えば`(\d)\1`と書くと，同じ数字が2回連続で続くようなパターンを検索できます．
（`\d`が数字1文字を表し，`\1`が最初の括弧の中身にマッチした文字列を表す）

<font color="OrangeRed">(っ´･ω･)っ</font><font color="IndianRed">【上記の「前方の括弧」順について，括弧が入れ子になっている時の順番ですが，[開き括弧が前方にあるものから順に通し番号が振られます](https://docs.python.org/ja/3/howto/regex.html?highlight=番号を決めるには、単に開き括弧を左から右へ数え上げます#grouping)．】</font>

<font color="RoyalBlue">【実習】 文字列`"11234556"`から同じ数字が2回連続で続くようなパターンを検索する．
</font>

```
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
for m in re.finditer(r"(\d)\1", "11234556"):
    print(m.group())
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
```

マッチした文字列の一部分を取り出したい場合も，この通し番号を活用することができます．  
`re.search()`や`re.finditer()`によって出力されたマッチングオブジェクトに対して `group()`，`start()`，`end()`などを呼ぶ際，引数に上記の通し番号を指定することで，括弧内の文字列についての情報を得ることができます．


<font color="RoyalBlue">【実習】 以下の文に対して2020から2029までの数字の左にある単語を検索する．

`Summer Olympic Games: Brisbane 2032, Los Angeles 2028, Paris 2024`
</font>

```
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
for m in re.finditer("(\w+)\s202[0-9]", text):
    print(m.group(1), m.start(1), m.end(1))
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
```

<font color="OrangeRed">(っ´･ω･)っ</font><font color="IndianRed">【上記のコードで，`\w+`は1文字以上の英数字を表します．括弧の中身に対して1という通し番号が振られているため，`group(1)`とすると，その中身の文字列を取り出すことができます．】</font>

<font color="OrangeRed">(っ´･ω･)っ</font><font color="IndianRed">【表示された文字列はオリンピックの開催地になっていません（開催地であればAngelesでなく，Los Angelesと表示すべきです）．下記のように正規表現を少し複雑にすることで，開催地を表示することができます．  
```
for m in re.finditer("((\w+ )*\w+) 202[0-9]", text):
    print(m.group(1), m.start(1), m.end(1))
```

】</font>

## 4-3. <font color="Teal">置換</font>

文字列の中から指定のパターンを検索し，それを他の文字に置換することができます．  
置換を行うには，[`re.sub()`](https://docs.python.org/ja/3/library/re.html#re.sub)を使用します．  
`re.sub(検索パターン, 置換後文字列, 対象の文書)` という風に指定すると，`対象の文書`を検索し，見つかった`検索パターン`を指定した`置換後文字列`に置換します．そして，置換後の文字列を出力します．

<font color="RoyalBlue">【実習】 文字列`"I am 43 yeas old."` の 数字部分を `x` に置換する．
</font>

```
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
re.sub(r"\d", "x", "I am 43 years old.")
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
```

置換後の文字列を関数を使って生成することもできます．これにより，より複雑な置換処理を行うことができます．  
それを行うには，`re.sub(検索パターン, 文字列生成関数, 対象の文書)` という風に指定します．  
`文字列生成関数` は検索パターンにマッチするマッチングオブジェクトを受け取り，文字列を返すようにします．

<font color="RoyalBlue">【実習】 文字列`"You got 65. I got 58."` の「`got 点数`」の部分を，点数が60点以上なら`"passed"`に置換し，そうでないなら`"failed"` に置換する．
</font>

```
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
def get_result_string(m):
    score = int(m.group(1))
    if score < 60:
        return "failed"
    else:
        return "passed"

re.sub(r"got (\d+)", get_result_string, "You got 65. I got 58.")
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
```

<font color="OrangeRed">(っ´･ω･)っ</font><font color="IndianRed">【関数内の処理の詳細ですが，`m.group(1)` は，マッチしたパターンから1番目の括弧`"(\d+)"`の中身を取り出す処理です．それを整数に変換後，点数に応じて置換後の文字列を変えています．】</font>

## 4-4. <font color="Teal">文字列の分割</font>

文字列を単語などに分割する際，区切り文字として，正規表現を指定することができます．  
`re.split(区切り文字, 対象の文書)` という風に指定すると，`区切り文字`によって`対象の文書`を分割し，リストに格納します．

空白文字として`\s+`（1文字以上の空白文字）を指定することで，スペースやタブなどが混在している文書であっても，適切に単語への分解ができるようになります．

<font color="RoyalBlue">【実習】 文字列`"The first step is always the hardest."` を空白文字で分割し，リストを作成する．</font>

```
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
re.split(r'\s+', "The first step is always the hardest.")
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
```

<font color="OrangeRed">(っ´･ω･)っ</font><font color="IndianRed">【区切り文字を指定する際，区切り文字に括弧をつけると（例えば，`(\s+）`のようにすると），区切り文字自体も分割後のリストに含まれるようになります．これは便利な場合もありますが，もし区切り文字をリストに含めたくないなら，区切り文字に括弧をつけないように注意しましょう．】</font>

<font color="OrangeRed">(っ´･ω･)っ</font><font color="IndianRed">【区切り文字として，`\W+`が使われることもあります．このようにすると，英数字（+アンダーバー）以外のすべての文字が区切り文字として扱われます．】</font>

## 4-5. <font color="Teal">日本語に対する正規表現</font>

絵文字などの特殊な文字を除き，日本語に対しても正規表現が使えるようです（2021年12月現在）．
* `\w` には，日本語のひらがな，カタカナ，漢字，全角アルファベット，全角数字も含まれている模様（記号は含まれていない模様）
* `\d` には，日本語の全角数字０-９も含まれている模様
* `\s` には，日本語の全角スペースも含まれている模様

<font color="RoyalBlue">【実習】 文字列`"昼食はラーメン、夕食はカレーライスだ。食事は大事。"` に対し「○食」（○は文字）というパターンを検索する．</font>

```
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
re.findall(r"\w食", "昼食はラーメン、夕食はカレーライスだ。食事は大事。")
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
```

<font color="OrangeRed">(っ´･ω･)っ</font><font color="IndianRed">【`昼食`，`夕食`にマッチするのに対し，`。食`にはマッチしていません．`。`が単語を構成する文字（`\w`）とはみなされないためです．】</font>

<font color="RoyalBlue">【実習】 文字列`"定食Ａは６２０円、定食Bは540円だ"` に対し，数字（全角含む）を検索し，その数字と文字列の中での位置を出力する．</font>

```
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
for m in re.finditer(r'\d+', "定食Ａは６２０円、定食Bは540円だ"):
    print(m.group(), m.start(), m.end())
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
```

<font color="OrangeRed">(っ´･ω･)っ</font><font color="IndianRed">【全角数字，半角数字ともに検索できています．また，位置についても全角文字の文字数のindexとなっています．】</font>

<font color="RoyalBlue">【実習】 文字列`"大阪府 豊中市　宮城県 仙台市"` に対し，空白文字（全角含む）で文字列を分割する．</font>

```
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
re.split(r'\s+', "大阪府 豊中市　宮城県 仙台市")
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
```

<font color="OrangeRed">(っ´･ω･)っ</font><font color="IndianRed">【全角スペース，半角スペースともに区切り文字として使われています．】</font>

### 5-1-4. <font color="Teal">テキストの加工</font>

テキストファイルに対して，文字列の置換を行ったり，不要な行を削除したりするには，`sed`コマンドが便利です．  
`sed`コマンドの処理結果は標準出力に出力されます．  
そのためパイプラインを利用して，その文字列を繰り返し編集することも可能です．  
処理結果をテキストファイルに出力するには，標準出力の内容をファイルにリダイレクトします．

#### <font color="Teal">指定文字列パターンの削除</font>

上記テキストファイルですが，本文の各行の先頭に余計なスペースが入っているので，まずはそれを消してみましょう．  
`sed "s/置換したい文字列パターン/置換後の文字列/"  入力文字列` とすると入力文字列内にある指定文字列の置換を行うことができます．  
ここでは行頭の空白文字を空文字（長さ0の文字列）に置換することにより，空白文字の削除を行うことにします．

<font color="RoyalBlue">【実習】 テキストファイルの中身について，行頭に入っているスペースを削除する</font>

```
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
!sed "s/^\s*//" ochazukeno_aji_utf8.txt
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
```

<font color="OrangeRed">(っ´･ω･)っ</font><font color="IndianRed">【`^\s*` は行頭に挿入されたスペース文字のうち最長のものにマッチします．それを空文字に置換することで，行頭のスペースを消すことができます．】</font>

#### <font color="Teal">指定行の削除</font>

次に本文の後ろに書かれている書籍情報（「底本」以下の内容）を削除しましょう．  
「底本」が書かれた行が18行めなので，16行め以降を削除すれば良さそうです．  
`sed "削除行d"  入力文字列` とすると入力文字列内の指定行を削除できます．  
また，削除行を `削除開始行，削除終了行` のように書くことで開始行から終了行まで一気に削除することができます．  
行指定として，テキストの最後の行を指定するには，`\$` と指定します．

<font color="RoyalBlue">【実習】 上記の処理に加え，テキストファイルの16行め以降を削除する</font>

```
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
!sed "s/^\s*//" ochazukeno_aji_utf8.txt | sed "16,\$d"
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
```

<font color="OrangeRed">(っ´･ω･)っ</font><font color="IndianRed">【`16,\$d` は16行めから最終行(`\$`)までを削除するという意味です．】</font>

<font color="RoyalBlue">【実習】 上記の処理に加え，テキストファイルの先頭3行も削除する</font>

```
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
!sed "s/^\s*//" ochazukeno_aji_utf8.txt | sed "16,\$d" | sed "1,3d"
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
```

<font color="OrangeRed">(っ´･ω･)っ</font><font color="IndianRed">【今回はこのような処理は不要ですが，文字数の多いテキストを扱う場合はこのような処理が必要になります．】</font>

## 5-3. <font color="Teal">HTMLファイルのダウンロードと前処理</font>

次に，HTMLタグ付きのWebページをダウンロードして，それに対して前処理を行います．  
BeautifulSoupというライブラリを使用して，HTMLコンテンツから所望の部分のみを切り出したり，不要なタグを削除したりする処理を行います．

### 5-3-1. <font color="Teal">HTMLファイルのダウンロード</font>

ここでは，青空文庫から宮沢賢治の[「注文の多い料理店」](https://www.aozora.gr.jp/cards/000081/card43754.html)という作品（著作権フリー）のWebページをダウンロードしてみます．  
[5-1-1](#scrollTo=QFeqG7q2pJzg)ではLinuxのwgetコマンドを利用して，マシンのストレージにファイルをダウンロードしましたが，ここではPythonプログラム内でファイルをダウンロードし，それをそのまま（ファイルとして保存せずに）使うことにします．  
そのために，[urllib.request](https://docs.python.org/ja/3/library/urllib.request.html)ライブラリを使用します．

その後，[BeautifulSoup](https://www.crummy.com/software/BeautifulSoup/bs4/doc/)ライブラリを利用して，ダウンロードしたファイルから不要なタグを削除し，本文のみを取り出します．  
BeautifulSoupはHTMLやXMLの解析を行うためのライブラリであり，HTMLファイル内で検索や編集を行うことができます．

<font color="RoyalBlue">【実習】 青空文庫から「注文の多い料理店」のHTMLファイルをダウンロードし，BeautifulSoupライブラリに読み込ませる</font>

```
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
from bs4 import BeautifulSoup
import urllib.request

url = "https://www.aozora.gr.jp/cards/000081/files/43754_17659.html"

with urllib.request.urlopen(url) as f:
    soup = BeautifulSoup(f)

print(soup)
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
```

### 5-3-2. <font color="Teal">HTMLコンテンツ内の検索</font>

BeautifulSoupを使うことで，様々な条件でHTMLファイル内を検索することができます．

<font color="RoyalBlue">【実習】 HTMLコンテンツから，`<title>`タグ(Webページのタイトル)
を検索し，最初に見つかったものを表示する</font>

```
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
soup.find("title")
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
```

<font color="OrangeRed">(っ´･ω･)っ</font><font color="IndianRed">【上記の代わりに `soup.title` と書くこともできます．】</font>


<font color="RoyalBlue">【実習】 HTMLコンテンツから，クラス（設定共通化のためのグループ）が `author` であるものをすべて検索し，リストとして出力する</font>

```
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
soup.find_all(class_='author')
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
```

<font color="OrangeRed">(っ´･ω･)っ</font><font color="IndianRed">【HTMLのクラスで検索するときは，キーワードとして，`class` でなく，`class_` と指定する必要があることに注意してください．[BeautifulSoupのドキュメント](https://www.crummy.com/software/BeautifulSoup/bs4/doc/#searching-by-css-class)によれば，`class`はPythonの予約語（Pythonでクラスを定義する際に使用する）なので使えず，代わりに `class_` を使うとのことです．】</font>


### 5-3-3. <font color="Teal">HTMLコンテンツの編集</font>

青空文庫からダウンロードしたHTMLコンテンツについて，本文の部分には，`class=main_text`が指定されているようです（[青空文庫の説明](https://www.aozora.gr.jp/annotation/extra.html)および上記のHTMLコンテンツ表示を参照）．  
よって，本文のみを取り出すには，`main_text`部分を取り出せば良いです．

また，HTMLコンテンツには，`<rt>`タグと`<rp>`タグを用いてふりがなの情報が挿入されています．  
しかし今回は不要なので，これらのタグ全体を消去することにします．  
検索で見つかったタグに対して，[`decompose()`](https://www.crummy.com/software/BeautifulSoup/bs4/doc/#decompose)メソッドを呼び出すことで，元のコンテンツからタグとタグの中身が消去されます．


<font color="RoyalBlue">【実習】 HTMLコンテンツから，`<rt>`タグと`<rp>`タグを検索し，これらすべてのタグとその中身を消去する</font>

```
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
for tag in soup.findAll(["rt", "rp"]):
    tag.decompose()
print(soup)
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
```

<font color="OrangeRed">(っ´･ω･)っ</font><font color="IndianRed">【青空文庫のHTMLコンテンツでは，`<rb>`タグで囲われた範囲に対して，`<rp>`タグと`<rt>`タグを使ってふりがなが挿入されています．また，それら全体が`<ruby>`タグで囲われています．今はふりがなだけを消去したいので，`<rp>`タグと`<rt>`タグのみを削除します．`<rb>`タグや`<ruby>`タグは残ったままになっていますが，この後タグの中身のテキストのみを取り出す処理を行います．】</font>


タグを含んだHTMLコンテンツから，タグを除いた中身の文章のみを取り出すには，[`get_text()`](https://www.crummy.com/software/BeautifulSoup/bs4/doc/#get-text)メソッドを使います．

<font color="RoyalBlue">【実習】 上記のHTMLコンテンツから，本文パート（`class=main_text`）部分を取り出し，その後，すべてのタグを取り除いて中身のコンテンツのみを残す</font>

```
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
kenji = soup.find(class_="main_text").get_text()
print(kenji)
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
```

<font color="OrangeRed">(っ´･ω･)っ</font><font color="IndianRed">【文章の先頭にはいくつかの改行が挿入されていますが，もし削除したい場合には，得られた文字列に対して`lstrip()`メソッドを適用します（`soup.find(class_="main_text").get_text().lstrip()`）．また，文章末尾の改行も削除する場合は`strip()`を使います．】</font>

### 5-3-5. <font color="Teal">ワードクラウドの改良</font>

上記のワードクラウドでは重要でない単語が大きく表示されてしまっています．  
そこでストップワードを設定することで，重要でない単語が表示されることを抑制します．

<font color="RoyalBlue">【実習】 [SlothLib](https://www.ieice.org/~de/DEWS/DEWS2007/pdf/c7-10.pdf)というライブラリで使用されているストップワードのリストを取得する</font>

```
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
url = "http://svn.sourceforge.jp/svnroot/slothlib/CSharp/Version1/SlothLib/NLP/Filter/StopWord/word/Japanese.txt"
with urllib.request.urlopen(url) as f:
    text = f.read().decode()

stopwords = []
for word in text.split('\r\n'):
    if len(word) < 1:
        continue
    stopwords.append(word)
print(stopwords)
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
```

<font color="RoyalBlue">【実習】 独自のストップワードリストを追加で作成した上で，ストップワードを除いた単語で分かち書きを作る</font>

```
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
words = []
my_stopwords = ["ます", "為る", "その", "其の", "其れ", "有る", "どう", "あり"]
mecab = MeCab.Tagger()
result = mecab.parse (kenji).splitlines()
for line in result:
    s = line.split("\t")
    if len(s) < 3:
        continue
    word = s[0]
    if word in stopwords:
        continue
    if word in my_stopwords:
        continue
    words.append(word)
sentence = " ".join(words)
print(sentence)
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
```

<font color="RoyalBlue">【実習】 上記で作成した分かち書きを使ってワードクラウドを作成する</font>

```
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
from wordcloud import WordCloud
import matplotlib.pyplot as plt
f_path = '/usr/share/fonts/opentype/ipafont-gothic/ipag.ttf'
wordcloud = WordCloud(background_color="white",
                      font_path=f_path, 
                      width=700, 
                      height=600).generate(sentence)
plt.figure(figsize=(12,10))
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#--#
```
